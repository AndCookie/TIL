## 전처리

```python
# 전처리
special_characters = ['(?)', ':)', ':(', '\n','-']

import string
punctuations = string.punctuation
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords') # 불용어 사전 다운로드
nltk.download('wordnet')
stop_words = stopwords.words('english')
from nltk.stem import WordNetLemmatizer, PorterStemmer
lemmatizer = WordNetLemmatizer()
# 형태학적 분석으로 단어의 원형 추출 : 정확한 결과 but 느린 속도
stemmer = PorterStemmer()
# 어미 제거로 어간 추출 : 빠른 속도 but 일부 단어의 원형을 제대로 추출하지 못함

def preprocess(text):      
    text = text.lower() # 소문자 변환
    text = re.sub(r'\.+', ' ', text)  # 중복 온점 제거
    text = text.replace('&', 'and') # & 제거
    
    text = re.sub(r'\d+', '', text)  # 숫자 제거
    text = re.sub(r'\s+', ' ', text)  # 중복 공백 제거
#     text = re.sub(r'(\w)(\1{1,})', r'\1', text) # 알파벳 3회이상 반복시 1개로
      
#    # 구두점 제거
    text = ''.join(char for char in text if char not in punctuations)
    # 불용어 제거
    text = ' '.join(word for word in text.split() if word not in stop_words)

#     # 표제어 추출
    # words = [lemmatizer.lemmatize(word) for word in text.split()]
#     # 어간 추출
#     words = [stemmer.stem(word) for word in text.split()]

#     return ' '.join(words)
    return text.strip()

def remove_space(text):
    for word in list(space['띄어쓰기 포함된 단어']):
        if word in text:
            new_word = word.replace(' ', '')
            text = text.replace(word, new_word)
    return text
```

```python
combined_review['전처리'] = [remove_space(preprocess(text)) for text in combined_review['제목_본문']]
combined_review['전처리'][:2]
```

```python
from ast import literal_eval
from typing import List

class Keywords:
    def __init__(self, voca_name: str):
        self.voca = pd.read_excel(voca_name)

    # kiwi에 단어 등록
    def add_word(self):
        for category in self.voca.columns:
            items = self.voca[category].dropna().apply(lambda x: literal_eval(x))

            for index, token in items.items():
                for form,tag in token.items():
                    kiwi.add_user_word(form, tag=tag, score=500)

      # 카테고리와 카테고리에 해당하는 값을 모두 출력
    def find_all(self, text: str) -> List[dict]:
        tokenized_text = kiwi.tokenize(text)
        tokens = [(token[0], token[1]) for token in tokenized_text]
        all = []
        for category in self.voca.columns:
            items = self.voca[category].dropna().apply(lambda x: literal_eval(x))
            for index, keyword in items.items():
                for kw in keyword.items():
                    if kw in tokens:
                        if kw[0] in list(space['띄어쓰기 없애기']):
                            for i, word in enumerate(space['띄어쓰기 없애기']):
                                if word == kw[0]:
                                    keyword = space.loc[i, '띄어쓰기 포함된 단어']
                                    category_dict = {category: keyword}
                                    all.append(category_dict)
                        else:
                            category_dict = {category: kw[0]}
                            all.append(category_dict)
        return all
```
