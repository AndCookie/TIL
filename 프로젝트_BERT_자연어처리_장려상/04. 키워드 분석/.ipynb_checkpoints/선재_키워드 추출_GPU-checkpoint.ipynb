## 전처리

```python
import torch

# 사용 가능한 GPU의 개수 확인
num_gpus = torch.cuda.device_count()

# GPU가 없는 경우에는 num_gpus를 0으로 설정
if num_gpus == 0:
    num_gpus = 1

print(f"Number of available GPUs: {num_gpus}")
```

```python
# 전처리
special_characters = ['(?)', ':)', ':(', '\n','-']

def preprocess(text):
    for special_ch in special_characters:
        # 특수문자 제거
        text = text.replace(special_ch, ' ')
        # 점 2개 이상 제거
        text = re.sub(r'\.{2,}', '', text)
        # 하나 이상의 공백 문자 찾고 공백 하나로
        text = re.sub(r'\s+', ' ', text)
        # 스페이스바 여러번 되어있는거 없애기
        text = ' '.join(text.split())
    return text

def remove_space(text):
    for word in list(space['띄어쓰기 포함된 단어']):
        if word in text:
            new_word = word.replace(' ', '')
            text = text.replace(word, new_word)
    return text
```

```python
combined_review['전처리'] = [remove_space(preprocess(text)) for text in combined_review['제목_본문']]
combined_review['전처리'][:2]
```

```python
def process_chunk(combined_review_chunk, result_queue, keywords):
    combined_review_chunk['키워드'] = combined_review_chunk['전처리'].apply(lambda x: keywords.find_all(x))
    result_queue.put(combined_review_chunk)
```

```python
# 전체 데이터를 청크 단위로 분할
chunk_size = len(combined_review) // num_gpus
chunks = [combined_review[i:i+chunk_size] for i in range(0, len(combined_review), chunk_size)]

# 결과를 저장할 큐 초기화
result_queue = mp.Queue()

# 각 프로세스마다 분할된 청크를 처리
processes = []
for i in range(num_gpus):
    p = mp.Process(target=process_chunk, args=(chunks[i], result_queue, keywords))
    p.start()
    processes.append(p)

# 모든 프로세스가 처리될 때까지 대기
for p in processes:
    p.join()

# 결과를 병합
combined_review = pd.concat([result_queue.get() for i in range(num_gpus)])
```

```python
from ast import literal_eval
from typing import List

class Keywords:
    def __init__(self, voca_name: str):
        self.voca = pd.read_excel(voca_name)

    # kiwi에 단어 등록
    def add_word(self):
        for category in self.voca.columns:
            items = self.voca[category].dropna().apply(lambda x: literal_eval(x))

            for index, token in items.items():
                for form,tag in token.items():
                    kiwi.add_user_word(form, tag=tag, score=500)

      # 카테고리와 카테고리에 해당하는 값을 모두 출력
    def find_all(self, text: str) -> List[dict]:
        tokenized_text = kiwi.tokenize(text)
        tokens = [(token[0], token[1]) for token in tokenized_text]
        all = []
        for category in self.voca.columns:
            items = self.voca[category].dropna().apply(lambda x: literal_eval(x))
            for index, keyword in items.items():
                for kw in keyword.items():
                    if kw in tokens:
                        if kw[0] in list(space['띄어쓰기 없애기']):
                            for i, word in enumerate(space['띄어쓰기 없애기']):
                                if word == kw[0]:
                                    keyword = space.loc[i, '띄어쓰기 포함된 단어']
                                    category_dict = {category: keyword}
                                    all.append(category_dict)
                        else:
                            category_dict = {category: kw[0]}
                            all.append(category_dict)
        return all
```
