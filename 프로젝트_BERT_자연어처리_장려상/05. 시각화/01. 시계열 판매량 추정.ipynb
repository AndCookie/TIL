# 데이터 로드

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글세팅
# 브랜드별 색상 팔레트 지정
colors = {'biobidet':'firebrick', 'brondell': 'coral', 'tushy':'gold', 'toto': 'lightgreen', 'ovedecors':'seagreen', 'alphabidet':'skyblue', 'kohler': 'royalblue', 'bidetmate':'mediumpurple', 'a_s': 'indigo', 'vovo':'black'}
# 채널별 색상 팔레트 지정
channel = {'costco':'black', 'amazon': 'darkgrey', 'build.com':'darkgoldenrod', 'walmart': 'gold', 'hd':'yellow', 'lowes':'orange'}

combined_review_origin = pd.read_excel('../04. 키워드 분석/keyword.xlsx')
combined_review_origin.rename(columns={'고유 제품명':'제품명'}, inplace=True)
combined_review_origin.info()
```

```python
# 카피본 떠놓고 시작하기
combined_review = combined_review_origin.copy()
combined_review.shape
```

```python
import ast

# 키워드 컬럼 내의 리스트를 문자열에서 리스트로 변환하여 저장
combined_review['키워드'] = combined_review['키워드'].apply(lambda x: ast.literal_eval(x) if len(x) > 2 else [])
# 리스트 내부의 딕셔너리를 문자열로 변환
combined_review['키워드'] = combined_review['키워드'].apply(lambda x: ', '.join([str(i) for i in x]))
# 문자열에서 [와 ]를 제거
combined_review['키워드'] = combined_review['키워드'].str.replace('[', '').str.replace(']', ',')
combined_review
```

```python
# 분기별/연도별로 날짜를 그룹화하고 리뷰 개수를 계산
combined_review['리뷰수'] = 1
combined_review['연도'] = combined_review['작성일'].dt.to_period("Y")
combined_review['분기'] = combined_review['작성일'].dt.to_period("Q")
combined_review
```

```python
target_products = ['Alpha JX', 'VB-6100SR', 'SW3074 C2', 'SW3056 S550', 'LT99', 'SE600', '3000 Series', 
                   'Classic 3.0', 'S1400', 'S1000', 'SW3036 K300', 'SW583 S350', 'SW3084 C5', 'C3-230',
                   'BB1000', 'C3-155', 'Basic 2.0', 'BB600', 'BB2000', 'BB1700']

# 제품명에 대한 리스트 요소가 target_products에 있는지 체크하는 함수
def check_product_name(product_name):
    return any(target_product in product_name for target_product in target_products)

# 제품명에서 target_products에 포함된 단어가 있는 데이터만 추출
target_reviews = combined_review[combined_review['제품명'].apply(check_product_name)]
target_reviews
```

```python
keyword_by_product = target_reviews.groupby('제품명')['키워드'].apply(lambda x: ' '.join(x)).reset_index()
keyword_by_product
```

```python
import ast
from collections import defaultdict

# 문자열을 딕셔너리로 변환하는 함수 정의
def str_to_dict(s):
    try:
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        return {}
    
result = {}
for product_name, keyword_str in zip(keyword_by_product['제품명'], keyword_by_product['키워드']):
    keyword_list = [str_to_dict(keyword.strip()) for keyword in keyword_str.split(',')]
    product_dict = {}
    for keyword_dict in keyword_list:
        for k, v in keyword_dict.items():
            product_dict[k] = product_dict.get(k, []) + [v]
    result[product_name] = product_dict
            
# 결과 딕셔너리 출력
for k, v in result.items():
    print(f"{k}: {v}")
            
# 결과 데이터프레임 생성
result_df = pd.DataFrame.from_dict(result, orient='index')
result_df.index.name = '제품명'
result_df = result_df.reset_index().rename(columns={'index': '키워드'})
result_df = result_df[['키워드'] + list(result_df.columns[1:])]

# 결과 출력
print(result_df)
```
