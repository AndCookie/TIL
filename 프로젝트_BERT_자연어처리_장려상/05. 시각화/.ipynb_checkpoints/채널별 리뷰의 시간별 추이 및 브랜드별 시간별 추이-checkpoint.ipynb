# 제품별 가중평균

```python
df_concat2=df_concat1.copy()
```

```python
count_=df_concat2.groupby(['회사명', '제품명'])['별점'].count().reset_index(name='별점 수')
count_
```

```python
mean_df=df_concat2.groupby(['회사명', '제품명'])['별점'].mean().reset_index(name='평균 별점')
```

```python
temp_df=pd.merge(count_,mean_df,on=(['회사명','제품명']))
temp_df
```

```python
temp_df['별점 총합']=temp_df['별점 수']*temp_df['평균 별점']
```

```python
def weighted_avg_log(df):
    # 로그 평균 적용
    df['로그전체'] = np.log10(df['별점 총합'] + 1)
    df['로그리뷰수'] = np.log10(df['별점 수'] + 1)
    
    # 가중평균 구하기
    total_weight = df['로그리뷰수'].sum()
    df['가중평균'] = (df['로그전체'] * df['로그리뷰수']).sum() / total_weight
    
    return df
```

```python
log_avg=temp_df.groupby(['회사명', '제품명']).apply(weighted_avg_log)
```

```python
log_avg.sort_values('가중평균',ascending=False).head(20)
```

```python
def weighted_average(df):
    # 별점과 리뷰 수를 모두 고려하여 가중평균 계산
    weighted_sum = (df['rating'] * np.log10(df['reviews'] + 1)).sum()
    total_weight = np.log10(df['reviews'] + 1).sum()
    
    # 리뷰 수가 0일 경우는 NaN 반환
    if total_weight == 0:
        return np.nan
    else:
        return weighted_sum / total_weight
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))

# 각 회사의 제품별로 별점 평균 구하기
df_avg = df_concat.groupby(['channel', '회사명', '제품명', pd.Grouper(key='작성일', freq='M')]).mean().reset_index()

# 시각화
for company in df_avg['회사명'].unique():
    df_company = df_avg[df_avg['회사명'] == company]
    for x, y in zip(df_company['작성일'], df_company['별점']):
        plt.text(x, y, str(y), ha='center', va='bottom')
    plt.plot('작성일', '별점', data=df_company, label=company)
    plt.title("제품별 별점 추이")
    plt.xlabel("작성일")
    plt.ylabel("별점")
    plt.tight_layout()
    plt.grid()
    
plt.legend(loc='best')
plt.show()
```

```python
df1['year'] = pd.to_datetime(df1['작성일']).dt.year
df1['quarter'] = pd.to_datetime(df1['작성일']).dt.quarter

# 피벗 테이블 생성
df_pivot = pd.pivot_table(df1, index=['year', 'quarter', '회사명', '제품명'], values='별점', aggfunc='mean').reset_index()

# 시각화
sns.set(style="darkgrid",font='Malgun Gothic')
companies = df_pivot['회사명'].unique()
products = df_pivot['제품명'].unique()

fig, axes = plt.subplots(nrows=len(companies), ncols=len(products), figsize=(20, 20))
fig.suptitle('년도별 분기별 제품별 회사별 평균 별점')

for i, company in enumerate(companies):
    for j, product in enumerate(products):
        data = df_pivot[(df_pivot['회사명'] == company) & (df_pivot['제품명'] == product)]
        sns.lineplot(x='quarter', y='별점', hue='year', data=data, ax=axes[i, j])
        axes[i, j].set_title(f"{company} - {product}")
        axes[i, j].set_ylim(0, 5)
        axes[i, j].set_xlabel('분기')
        axes[i, j].set_ylabel('평균 별점')

plt.tight_layout()
plt.show()
```
