# 데이터 로드

```python
import os
import pandas as pd

file_path = '../01-1. Raw Data (Bidet)/' # 파일 경로
file_extension = '.xlsx' # 파일 확장자

file_list = [f for f in os.listdir(file_path) if f.endswith('_review_product' + file_extension)]
file_list
```

```python
data_frames = []
for file_name in file_list:
    channel = file_name.split('_')[0]
    df = pd.read_excel(file_path + file_name)
    df['채널명'] = channel
    data_frames.append(df)
```

# 데이터 통합

```python
combined_review = pd.concat(data_frames, ignore_index=True)
combined_review.head()
```
