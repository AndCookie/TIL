## 회사명 추출

```python
com=[]
for i in build_review['제품명']:
    i.split(' ')
    a=i.lower().split(' ')[0]
    com.append(a)
print(com)
```

```python
build_review['회사명'] = com
build_review['회사명'].unique()
```

```python
build_review['회사명'] = build_review['회사명'].str.replace('american', 'a_s')
build_review['회사명'].unique()
```

## 제목_본문 통합

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1==col_2 :
        result=col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

build_review["제목_본문"] = build_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```

```python
build_review = build_review[['회사명', '작성일', '별점', '제목_본문']]
build_review
```
