# QnA

```python
import numpy as np
import pandas as pd

amazon_QnA= pd.read_excel('./Sample_Amazon(US)_QnA_221209.xlsx')
amazon_QnA.shape
amazon_QnA.head()
```

```python
amazon_QnA.shape
```

```python
amazon_QnA.info() #답변일자, 답변
```

## copy 생성

```python
amazon = amazon_QnA.copy()
amazon.head(20)
```

```python
amazon['제품명'].head(30)
```

```python
com=[]
for i in amazon['제품명']:
    i.split(' ')
    a=i.split(' ')[0]
    com.append(a)
print(com)
```

```python
amazon['회사명'] = com
```

```python
amazon['회사명'].unique()
```

```python
amazon['회사명'] = amazon['회사명'].str.replace('ALPHA', 'alphabidet')
amazon['회사명'] = amazon['회사명'].str.replace('Bio', 'BioBidet')
amazon['회사명'] = amazon['회사명'].str.replace('TUSHY', 'Tushy')
amazon['회사명'] = amazon['회사명'].str.replace('BioBidetBidet', 'BioBidet')
amazon['회사명'] =amazon['회사명'].str.lower()
amazon['회사명'].unique()
```

## 중복확인

```python
amazon[amazon.duplicated()]
```

```python
amazon.columns
```

```python
amazon=amazon.drop(['추출 날짜', '수집채널','QnA URL','제품상세URL'], axis=1)
amazon
```

```python
amazon=amazon[['회사명', '제품명', '질문일자','질문','답변일자','답변']]
```
