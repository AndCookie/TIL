```python
company_list=[]

for url in lowes_review['제품상세URL']:
    company_list.append(url.lower().split('/pd/')[1].split('-')[0])
print(set(company_list))
```

```python
lowes_review['회사명'] = company_list
lowes_review['회사명'].unique()
```

```python
lowes_review['회사명'] = lowes_review['회사명'].str.replace('bio', 'biobidet').replace('american', 'a_s')
lowes_review['회사명'].unique()
```

## 제목_본문 통합

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1==col_2 :
        result=col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

lowes_review["제목_본문"] = lowes_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```

```python
lowes_review=lowes_review[['회사명', '작성일', '별점', '제목_본문']]
lowes_review
```
