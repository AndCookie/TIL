## 날짜 계산

```python
date = costco_review[['추출 날짜','작성일']]
date['작성일'].unique()
```

```python
import datetime

# 주어진 날짜와 상대적인 날짜 표현
date_str = date['추출 날짜']
relative_date = date['작성일']
date_list = []

# 주어진 날짜 문자열을 datetime 객체로 변환
for i in range(len(date_str)):
    date_obj = datetime.datetime.strptime(date_str[i], '%Y.%m.%d')

    # 상대적인 날짜 표현을 해석하여 해당 날짜를 구함
    if 'years' in relative_date[i]:
        years_ago = int(relative_date[i].split()[0])
        if years_ago == 0:
            years_ago += 1
        date_obj = date_obj - datetime.timedelta(days=years_ago*365)
    elif 'year' in relative_date[i]:
        date_obj = date_obj - datetime.timedelta(days=365)
    elif 'months' in relative_date[i]:
        months_ago = int(relative_date[i].split()[0])
        if months_ago == 0:
            months_ago += 1
        date_obj = date_obj - datetime.timedelta(days=months_ago*30)
    elif 'weeks' in relative_date[i]:
        weeks_ago = int(relative_date[i].split()[0])
        if weeks_ago == 0:
            weeks_ago += 1
        date_obj = date_obj - datetime.timedelta(weeks=weeks_ago)
    elif 'days' in relative_date[i]:
        days_ago = int(relative_date[i].split()[0])
        if days_ago == 0:
            days_ago += 1
        date_obj = date_obj - datetime.timedelta(days=days_ago)
    elif 'hours' in relative_date[i]:
        hours_ago = int(relative_date[i].split()[0])
        if hours_ago == 0:
            hours_ago += 1
        date_obj = date_obj - datetime.timedelta(hours=hours_ago)

    # 변환된 날짜를 리스트에 추가
    date_list.append(date_obj)

# 결과 출력
for date in date_list:
    print(date.strftime('%Y.%m.%d'))
```

```python
costco_review['작성일'] = date_list
costco_review
```

# 고유 제품명 / 형태

```python
#고유 제품명 들어있는 데이터 불러오기
import pandas as pd

excel_file = './Sample_Costco_Review_221205.xlsx'  # 엑셀 파일 경로 및 이름
sheet_name = 2  # 시트 번호 (0부터 시작)

df = pd.read_excel(excel_file, sheet_name=sheet_name)
```

```python
def add_column_b_from_a(a, b, a_col, b_col, new_col,new_col2):
    """
    a_col과 b_col 컬럼을 비교하여 같으면 a의 B 컬럼 값을 b의 새로운 new_col 컬럼에 할당하는 함수

    Parameters:
    a (pandas.DataFrame): 비교할 a 데이터프레임
    b (pandas.DataFrame): 값을 할당할 b 데이터프레임
    a_col (str): 비교할 a 데이터프레임의 컬럼 이름
    b_col (str): 비교할 b 데이터프레임의 컬럼 이름
    new_col (str): 값을 할당할 b 데이터프레임의 새로운 컬럼 이름

    Returns:
    pandas.DataFrame: 값을 할당한 b 데이터프레임
    """
    # b 데이터프레임에 새로운 컬럼 추가
    b[new_col] = None
    b[new_col2] =None

    # a와 b의 컬럼 비교하여 값을 할당
    for idx, row_b in b.iterrows():
        a_val = a.loc[a[a_col] == row_b[b_col], "Unnamed: 2"].values
        if len(a_val) > 0:
            b.loc[idx, new_col] = a_val[0]
    for idx, row_b in b.iterrows():
        a_val = a.loc[a[a_col] == row_b[b_col], "Unnamed: 3"].values
        if len(a_val) > 0:
            b.loc[idx, new_col2] = a_val[0]
    return b

```

## 제목_본문 통합

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1==col_2 :
        result=col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

costco_review["제목_본문"] = costco_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```

```python
costco_review['제목_본문']=costco_review['제목_본문'].str.replace("‘", "’").str.replace("’", "’").str.replace("'", "’")
```
