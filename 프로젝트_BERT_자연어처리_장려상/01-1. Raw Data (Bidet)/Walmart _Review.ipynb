# 데이터 전처리

## 제목_본문 통합

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1==col_2 :
        result=col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result


Walmart_Review["제목_본문"] = Walmart_Review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```

```python
Walmart_Review
```

## 컬럼 제거

```python
Walmart_Review = Walmart_Review.drop(['제목', '본문', '이미지URL' ,'리뷰URL'], axis=1)

Walmart_Review.columns
```

## NA 제거

```python
idx = Walmart_Review[Walmart_Review['제목_본문']=='nan'].index
Walmart_Review=Walmart_Review.drop(idx)
```

## 회사명 추출


```python
Walmart_Review['제품명']
```

```python
Walmart_Review['제품명'].str.contains('Bio|Brondell',case=False).sum()
```

```python
com=[]
for i in Walmart_Review['제품명']:
    i.split(' ')
    a=i.split(' ')[2]
    com.append(a.lower())
print(com)

Walmart_Review['회사명']=com
Walmart_Review['회사명'].unique()
```

```python
Walmart_Review['회사명'] = Walmart_Review['회사명'].replace('bio', 'biobidet').replace('alpha', 'alphabidet')
Walmart_Review['회사명'].unique()
```

```python
Walmart_Review = Walmart_Review[['회사명', '작성자', '작성일', '별점', '제목_본문']]
Walmart_Review.head()
```
