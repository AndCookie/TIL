```python
def add_column_b_from_a(a, b, a_col, b_col, new_col,new_col2):
    """
    a_col과 b_col 컬럼을 비교하여 같으면 a의 B 컬럼 값을 b의 새로운 new_col 컬럼에 할당하는 함수

    Parameters:
    a (pandas.DataFrame): 비교할 a 데이터프레임
    b (pandas.DataFrame): 값을 할당할 b 데이터프레임
    a_col (str): 비교할 a 데이터프레임의 컬럼 이름
    b_col (str): 비교할 b 데이터프레임의 컬럼 이름
    new_col (str): 값을 할당할 b 데이터프레임의 새로운 컬럼 이름

    Returns:
    pandas.DataFrame: 값을 할당한 b 데이터프레임
    """
    # b 데이터프레임에 새로운 컬럼 추가
    b[new_col] = None
    b[new_col2] =None

    # a와 b의 컬럼 비교하여 값을 할당
    for idx, row_b in b.iterrows():
        a_val = a.loc[a[a_col] == row_b[b_col], "Unnamed: 2"].values
        if len(a_val) > 0:
            b.loc[idx, new_col] = a_val[0]
    for idx, row_b in b.iterrows():
        a_val = a.loc[a[a_col] == row_b[b_col], "Unnamed: 3"].values
        if len(a_val) > 0:
            b.loc[idx, new_col2] = a_val[0]
    return b

```

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1 == col_2 :
        result = col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

amazon_review["제목_본문"] = amazon_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```
