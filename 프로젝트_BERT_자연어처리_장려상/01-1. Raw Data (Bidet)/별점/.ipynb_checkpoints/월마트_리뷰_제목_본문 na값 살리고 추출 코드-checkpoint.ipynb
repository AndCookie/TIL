# 제목_본문 통합

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1==col_2:
        result=col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

walmart_review['제목_본문'] = walmart_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']),axis=1)
```

```python
walmart_review['제목_본문']=walmart_review['제목_본문'].str.replace("‘", "’").str.replace("’", "’").str.replace("'", "’")
```

```python
walmart_review=walmart_review[['회사명','제품명','작성일','별점']]
walmart_review
```
