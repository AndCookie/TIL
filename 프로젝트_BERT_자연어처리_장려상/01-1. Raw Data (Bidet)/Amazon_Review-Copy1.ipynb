# 데이터 전처리

## 회사명 추출

```python
com=[]
for i in amazon_review['제품명']:
    i.split(' ')
    a=i.lower().split(' ')[0]
    com.append(a)
print(com)
```

```python
amazon_review['회사명'] = com
amazon_review['회사명'].unique()
```

```python
amazon_review['회사명'] = amazon_review['회사명'].str.replace('bio', 'biobidet').replace('alpha', 'alphabidet').replace('biobidetbidet', 'biobidet')
amazon_review['회사명'].unique()
```

```python
amazon_review.columns
```

## 제목_본문 통합

```python
amazon_review = amazon_review[['회사명', '제품명', '작성자', '작성일', '별점', '제목', '본문']]
amazon_review[:5]
```

```python
def combine_2nd_columns(col_1, col_2):
    if pd.isna(col_1):
        result=col_2
    elif col_1 == col_2 :
        result = col_1
    else:
        result = col_1
        if not pd.isna(col_2):
            result += "." + str(col_2)
    return result

amazon_review["제목_본문"] = amazon_review.apply(lambda x: combine_2nd_columns(x['제목'], x['본문']), axis=1)
```

```python
amazon_review = amazon_review[['회사명', '제품명', '작성자', '작성일', '별점', '제목_본문']]
amazon_review[:5]
```

## 데이터 타입 변환

```python
amazon_review['작성일']=pd.to_datetime(amazon_review['작성일'])
```
