# 전처리

```python
accuracy_df = pd.DataFrame(columns=['Train Data', 'Test Data', 'Base (기본 전처리 = 소문자화, 구두점·숫자 제거) + 불용어 제거', 'Base + 정규화 (표제어 추출 WordNetLemmatizer)', 'Base + 정규화 + 하이퍼 파라미터 튜닝'])
accuracy_df.loc['GB1'] = ['Open', 'Bidet', '71.65%', '78.47%', '84.59%']
accuracy_df.loc['GB2'] = ['Bidet+Open', 'Bidet', '83.96%', '91.18%', '93.36%']
accuracy_df.loc['SVM'] = ['Bidet+Open', 'Bidet', '94.56%', '96.52%', '96.55%']

accuracy_df.style.set_properties(**{'text-align': 'center'}).set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
accuracy_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

# 데이터프레임에서 정확도 데이터만 추출합니다.
data = accuracy_df.iloc[:, 2:].apply(lambda x: x.str.rstrip('%').astype('float') / 100)

# heatmap을 그립니다.
plt.figure(figsize=(10, 5))
sns.heatmap(data, annot=True, cmap='Blues')
plt.title('Accuracy Comparison')
plt.show()
```

```python
accuracy_df = pd.DataFrame(columns=['분석 방법', 'Train Data', 'Test Data', 'Base (기본 전처리 = 소문자화, 구두점·숫자 제거) + 불용어 제거', 'Base + 정규화 (표제어 추출 WordNetLemmatizer)', 'Base + 정규화 + 하이퍼 파라미터 튜닝'])
accuracy_df.loc['1'] = ['Gradient Boosting', 'Open', 'Bidet', '71.65%', '78.47%', '84.59%']
accuracy_df.loc['2'] = ['Gradient Boosting', 'Bidet+Open', 'Bidet', '83.96%', '91.18%', '93.36%']
accuracy_df.loc['3'] = ['SVM', 'Bidet+Open', 'Bidet', '94.56%', '96.52%', '96.55%']
# accuracy_df.loc['4'] = ['BERT', 'Bidet+Open', 'Bidet', '98.56%', '98.56%', '98.56%']

accuracy_df.style.set_properties(**{'text-align': 'center'}).set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
accuracy_df
```

```python
import matplotlib.pyplot as plt

# 막대 그래프를 그리기 위해 데이터프레임의 컬럼과 데이터를 분리합니다.
method = accuracy_df['분석 방법'].values
train_data = accuracy_df['Train Data'].values
test_data = accuracy_df['Test Data'].values
preprocessing = accuracy_df[['Base (기본 전처리 = 소문자화, 구두점·숫자 제거) + 불용어 제거', 'Base + 정규화 (표제어 추출 WordNetLemmatizer)', 'Base + 정규화 + 하이퍼 파라미터 튜닝']].apply(lambda x: x.str.rstrip('%')).astype(float).values
bert = [98.56]

# 그래프의 크기를 설정합니다.
fig, ax = plt.subplots(figsize=(16,12))

# 각 막대의 위치를 설정합니다.
bar_width = 0.2
opacity = 0.8
index = np.arange(len(train_data))

# 각 막대를 그립니다.
rects1 = ax.bar(index, preprocessing[:,0], bar_width,
                alpha=opacity,
                color='pink',
                label='Base (기본 전처리 = 소문자화, 구두점·숫자 제거) + 불용어 제거')

rects2 = ax.bar(index + bar_width, preprocessing[:,1], bar_width,
                alpha=opacity,
                color='palevioletred',
                label='Base + 정규화 (표제어 추출 WordNetLemmatizer)')

rects3 = ax.bar(index + 2*bar_width, preprocessing[:,2], bar_width,
                alpha=opacity,
                color='crimson',
                label='Base + 정규화 + 하이퍼 파라미터 튜닝')

rects4 = ax.bar(index + 3*bar_width, bert, bar_width,
                alpha=opacity,
                color='blue',
                label='딥러닝 BERT')

# rects5 = ax.bar(index + 4*bar_width, preprocessing[:,4], bar_width,
#                 alpha=opacity,
#                 color='crimson',
#                 label='Base + 소문자화, 구두점 제거, 불용어 제거')

# 축과 그래프의 레이블을 설정합니다.
ax.set_xlabel('머신러닝 분석 기법 : 학습 데이터 종류', fontsize=15)
ax.set_ylabel('정확도', fontsize=15)
ax.set_title('Accuracy by Data and Preprocessing', fontsize=20)
ax.set_xticks(index + bar_width)
ax.set_xticklabels(method + ' : ' + train_data, fontsize=12, rotation=0)
ax.legend(fontsize=12, loc='upper left')

# y축 범위를 50~100%로 설정합니다.
ax.set_ylim(70, 100)

# 각 막대에 대한 값을 표시합니다.
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., height + 0.5,
                '{:.2f}%'.format(height),
                ha='center', va='bottom', fontsize=12)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
# autolabel(rects5)

plt.tight_layout()

# 그래프를 이미지 파일로 저장합니다.
filename = 'comparison.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')

plt.show()
```
