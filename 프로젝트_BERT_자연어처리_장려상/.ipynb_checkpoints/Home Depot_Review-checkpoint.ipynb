```python
# 카피본 떠놓고 시작하기
hd_review = HomeDepot_Review_origin.copy()
hd_review.info()
```

# 데이터 전처리

## 컬럼 삭제

```python
hd_review = hd_review.drop(['추출 날짜', '수집채널','작성자' ,'이미지URL' ,'리뷰URL'], axis=1)

hd_review.columns
```

## 데이터 타입 변환

```python
hd_review['작성일']=pd.to_datetime(hd_review['작성일'])
```

```python
hd_review.info()
```

## 회사명 추출

```python
hd_review['제품상세URL']
```

```python
hd_review['제품상세URL'].str.contains('toto|brondell|alpha-bidet|biobidet', case=False).sum()
```

```python
company_list=[]

for url in hd_review['제품상세URL']:
    company_list.append(url.lower().split('/p/')[1].split('-')[0])
print(set(company_list))
```

```python
hd_review['회사명'] = company_list
hd_review['회사명']
```

```python
hd_review['회사명']
```

```python
hd_review = hd_review[['회사명', '제품명', '작성일', '별점', '제목', '본문']]
hd_review.head()
```

```python
hd_review['회사명'].unique()
```

```python
hd_review['회사명'] = hd_review['회사명'].str.replace('alpha', 'alphabidet')
hd_review['회사명'].unique()
```
