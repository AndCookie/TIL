# 전처리

## 한글 필터링
- re.sub(pattern, repl, string)
- string에서 pattern을 찾아서 repl로 치환한다.

```python
import re

def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub('', text)
    return result

ko_text = []
for r in df['review']:
    pre = text_cleaning(r)
    ko_text.append(pre)
    
df['ko_text'] = ko_text
df = df[df['ko_text'].str.len()>0]

print(df.shape) # (898, 4)
display(df)
```

```python
df.index = np.arange(0, len(df.index))
df.index
```

```python
print(df.shape) # (898, 4)
display(df)
```

```python
df = df[['score', 'y', 'ko_text']]
df.head()
```

## 형태소/품사 추출

```python
# pip install konlpy
! pip list
```

```python
from konlpy.tag import Okt

def get_pos(x):
    tagger = Okt()
    pos = tagger.pos(x)
    pos = ['{}/{}'.format(word, tag) for word, tag in pos]
    return pos

result = get_pos(df['ko_text'].values[0])
print(result)
```

## CounterVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# tokenizer : 텍스트 데이터의 전처리 방식 입력
index_vectorizer = CountVectorizer(tokenizer = lambda x:get_pos(x))
X = index_vectorizer.fit_transform(df['ko_text'].tolist())
print(X.shape) # (898, 5281)
# print(X)
```

```python
index_vectorizer.vocabulary_
```

```python
print(df['ko_text'][0])
print(X[0])
```

## TF-IDF
- Term Frequency - Inverse Documnet Frequency
- 다른 문서들에서는 등장하지 않았지만, 현재 문서에서는 많이 등장하는 단어
- 해당 단어가 현재 문서에서 얼마나 중요한지 피처로 나타낼 수 있는 방법
- 단어 빈도를 나타내는 TF * 문서 빈도를 나타내는 DF의 역수인 IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_vectorizer = TfidfTransformer()
X = tfidf_vectorizer.fit_transform(X)

print(X.shape) # (898, 5281)
print(X[0])
```

# 긍정/부정 리뷰 분류

## 학습/평가용 분리

```python
from sklearn.model_selection import train_test_split

y = df['y']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(x_train.shape)
print(x_test.shape)
```

## 로지스틱 회귀모델

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr = LogisticRegression(random_state=0)
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
y_pred_probability = lr.predict_proba(x_test)[:,1]

print('accuracy: %.2f' % accuracy_score(y_test, y_pred))
print('Precision: %.3f' % precision_score(y_test, y_pred))
print('Recall: %.3f' % recall_score(y_test, y_pred))
print('F1: %.3f' % f1_score(y_test, y_pred))
```

## 클래스 불균형

```python
# 클래스 불균형
from sklearn.metrics import confusion_matrix

cf = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(cf)
```

## 1:1 랜덤 샘플링

```python
positive_random_idx = df[df['y']==1].sample(50, random_state=20).index.tolist()
negative_random_idx = df[df['y']==0].sample(50, random_state=20).index.tolist()


random_idx = positive_random_idx + negative_random_idx

sample_X = X[random_idx, :]
y = df['y'][random_idx]

x_train, x_test, y_train, y_test = train_test_split(sample_X, y, test_size=0.3)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr = LogisticRegression(random_state=2)
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
y_pred_probability = lr.predict_proba(x_test)[:,1]

print('accuracy: %.2f' % accuracy_score(y_test, y_pred))
print('Precision: %.3f' % precision_score(y_test, y_pred))
print('Recall: %.3f' % recall_score(y_test, y_pred))
print('F1: %.3f' % f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cf = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(cf)
```

## 품사별 영향력 상위 Top_10
- 명사 / 형용사 / 동사

```python
noun_list = []
adjective_list = []
verb_list = []

for coef in coef_pos_index[:100]:
    pos_category = invert_index_vectorizer[coef[1]].split('/')[1]
    if pos_category == 'Noun':
        noun_list.append((invert_index_vectorizer[coef[1]], coef[0]))
    elif pos_category == 'Adjective':
        adjective_list.append((invert_index_vectorizer[coef[1]], coef[0]))
    elif pos_category == 'Verb':
        verb_list.append((invert_index_vectorizer[coef[1]], coef[0]))

print(noun_list[:15])
print('-'*15)
print(adjective_list[:15])
print('-'*15)
print(verb_list[:15])
```
