# 선 도표 (line plot) 그리기

## 기본적인 그래프 그리기
- plt.plot([x], y, [fmt(포맷팅)], [x2], y2, [fmt2], ...)
    - fmt : 색상, 마커, 선 스타일 등 기본 서식을 설정하는 것
- 하나의 plot에 여러 개의 좌표를 설정하면 복수 개의 그래프를 그릴 수 있다.

### y값

```python
plt.plot([0,1,2,3])
plt.show()
# [0,1,2,3] : y값 // 생략시 자동으로 y값으로 인식
# 마커 미지정 : 선으로 그려짐
```

- 위 코드에서 plot() 함수에 전달된 리스트는 **y축의 값으로 설정**된다.
- x 축의 값은 0부터 시작해서 1의 간격으로 자동 설정된다.

```python
plt.plot([0,1,2,3], 'o')
plt.show()
# 마커를 점o 표시기로 지정한 것
```

```python
plt.plot([0,1,2,3], 'x')
plt.show()
# 마커를 x 표시기로 지정한 것
```

## 스타일
- plot() 함수에 다양한 키워드 인수를 전달하여 스타일을 바꿀 수 있다

### 문자열로 지정하기

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2, 
        color='red',
        linestyle='dashed')
plt.show()
# 빨간색의 점선으로 이루어진 line plot
```

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2, 
        color='red',
        linestyle='dotted')
plt.show()
# 빨간색의 점으로 이루어진 line plot
```

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2, 
        color='red',
        linestyle='dotted',
        linewidth=2.5,
        marker='o',
        markeredgecolor='green',
        markerfacecolor='yellow',
        markersize=9.3)
plt.show()
# 빨간색의 2.5두께의 점선으로 이루어진 line plot
# marker : 각각의 데이터에 표시할 기호
```

### 상징문자로 지정하기
- 사전에 정의되어 있는 상징문자를 이용하여 스타일 지정


```python
x = np.linspace(0, 2*np.pi, 20)
# 주어진 데이터를 균등하게 분할하기
# 10 -> 20 분할 개수가 늘어날수록 유연한 데이터가 생성됨

plt.plot(x, np.sin(x), 'ro--', x, np.cos(x), 'g^-.', x, -np.sin(x), 'bx:' )
# 하나의 plot 내에 여러 개의 좌표를 설정하면 -> 여러 그래프 그릴 수 있다
# sin 그래프 : 빨간색 동그라미 & dashed 선으로
# cos 그래프 : 초록색 삼각형 & dash-dot 선으로
# 반전된 sin : 파란색 엑스 & dotted 선으로

plt.xlim(0, 2*np.pi)
# axis는 x축과 y축의 범위를 동시에 지정
# lim은 따로따로 지정

plt.show()
```

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2, 
        color='red',
        linestyle='dotted',
        linewidth=2.5,
        marker='o',
        markeredgecolor='green',
        markerfacecolor='yellow',
        markersize=9.3)
plt.show()
```

## 범례 표시하기
- plot() 함수에 label 인자를 설정함으로써 그래프에 이름을 지정할 수 있다.
- 이렇게 지정된 이름은 legend() 함수를 통해 범례로써 표시할 수 있다.
- loc 파라미터에 다음 문자열 값을 지정하면 해당 위치에 범례가 표시된다.(location: 우측 상단, 가운데 등)
<pre>
'best' : 그래프와 범례가 겹치지 않는 선에서 최적의 위치에 배치하기
'right'
'left'
'center'
'upper right'
'upper left'
'upper center'
'lower right'
'lower left'
'lower center'
'center right'
'center left'
</pre>

```python
a = np.arange(0,3,.02) # 0부터 3까지 0.02 간격으로
c = np.exp(a) # exp : 제곱근 함수
d = c[::-1]

# 하나의 그래프 안에 3개의 선 그래프가 그려진다.
plt.plot(a,c,'r--', label='모델 길이')
# 빨간색 dashed 그래프. 이름은 'Model length'
plt.plot(a,d,'b:', label='데이터 길이')
# 파란색 dotted 그래프. 이름은 'Data length'
plt.plot(a,c+d, 'g', label='전체 메세지 길이')
# 초록색 그래프. 이름은 'Total message length'

# 범례 표시하기
plt.legend(loc='best', shadow=True)
# 유니코드 설정해줘야 한글 범례 표시 가능
# 상단 가운데에 입체감 있게 그림자 표현

plt.show()
```

# 서프 플롯 그리기

## subplot
- subplot() : 하나의 figure에 여러 개의 그래프를 모두 그릴 수 있다.
- 주요 인자 : nrow, ncols, index를 지정하는 3개의 정수를 이용하여 서브플롯의 위치를 설정

```python
x1 = np.linspace(0.0, 5.0, 50) # 0부터 5까지를 50개로 균등 분할
x2 = np.linspace(0.0, 2.0, 50) # 0부터 2까지를 50개로 균등 분할

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)


# 사이즈 조정
plt.figure(figsize=(10,5))


plt.subplot(2,1,1) # nrow=2, ncols=1, index=1 # 2행 1열 형태인 2개의 그래프 중 첫 번째 해당하는 그림이다
plt.plot(x1, y1, 'o-')
plt.title('1st Graph') #오슬로의 감소 그래프?
plt.xlabel('times(s)')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2) # nrow=2, ncols=1, index=2 # 2행 1열 형태인 2개의 그래프 중 두 번째 해당하는 그림이다
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('times(s)')
plt.ylabel('Undamped')


# 서브플롯의 레이아웃 자동 맞춤 (겹치기 X)
plt.tight_layout()


plt.show()
```

## subplots

- subplots() : 서브플롯을 표시
- 표시할 플롯의 개수를 설정한 다음 반환되는 객체를 이용해서 각 플롯의 데이터를 설정한다.
- 주요 인수: 
    - nrows, ncols: 서브플롯의 행과 열의 개수 지정
    - sharex, sharey: boolean 값 (DEFAULT : False)으로 x, y축의 공유제어 속성

```python
fig, axes = plt.subplots(1,4, sharey=True)
# 1행 4열짜리 그림을 그리고, y축의 값은 공유하겠다


# 사이즈 조정시 plt.figure(figsize=)가 아닌 fig.set_size_inches()
fig.set_size_inches((16,4))


# 산점도
# 1행 4열이므로 axes에 하나의 값만 전달했으나 2행 3열이라면 축의 값 전달시 행,열 모두 명시해야한다
# alpha : 투명 농도
axes[0].scatter(np.random.rand(100), np.random.rand(100), c='cyan', edgecolor='red', alpha=0.5)
axes[1].scatter(np.random.rand(100), np.random.rand(100), c='magenta', edgecolor='green', alpha=0.5)
axes[2].scatter(np.random.rand(100), np.random.rand(100), c='yellow', edgecolor='blue', alpha=0.5)
axes[3].scatter(np.random.rand(100), np.random.rand(100), c='white', edgecolor='black', alpha=0.5)


# 축 객체 이용시 title이 아닌 set_title 함수 사용
axes[0].set_title('subplot 0')
axes[1].set_title('subplot 1')
axes[2].set_title('subplot 2')
axes[3].set_title('subplot 3')


plt.show()
```

```python
x = np.arange(0,10,0.01)

fig, axes = plt.subplots(2,2)
# 2행 2열짜리 그림


# 축의 값 전달시 행,열 모두 명시해야한다
axes[0][0].plot(x, np.sin(x))
axes[0][1].plot(x, np.cos(x))
axes[1][0].plot(x, np.tanh(x))
axes[1][1].plot(x, np.sin(x)**2)

# 축 객체 이용시 title이 아닌 set_title 함수 사용
axes[0][0].set_title('subplot 0')
axes[0][1].set_title('subplot 1')
axes[1][0].set_title('subplot 2')
axes[1][1].set_title('subplot 3')


plt.show()
```

# 막대 그래프 (bar char) 그리기

## bar()
- bar() : 막대 그래프 생성

```python
x = np.arange(0,20)
y = x *2


# 막대에 눈금 표시하기 : plt.xticks()
x_tics = np.arange(len(x)) # x의 개수만큼 눈금 생성
plt.bar(x, y, label='Blue Bar', color='blue', edgecolor='yellow', linewidth=3)

plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
plt.xticks(x_tics) # x축의 눈금 지정
plt.legend() # 범례 표시 # DEFAULT : BEST
plt.show()
```

## barh()
- barh() : 수평 막대 그래프 생성

```python
x = np.arange(0,20)
y = x *2


# 막대 눈금 표시
y_tics = np.arange(len(x)) # x의 개수만큼 눈금 생성


# 수평 막대 그래프를 그린다고 해서, x와 y 값의 위치가 바뀌는 것은 아니다!!
plt.barh(x, y, label='Blue Bar', color='blue', edgecolor='yellow', linewidth=3)

plt.xlabel('bar height')
plt.ylabel('bar number')
plt.title('H-Bar Chart Example')
plt.yticks(y_tics) # y축의 눈금 지정
plt.legend() # 범례 표시 # DEFAULT : BEST
plt.show()
```

- matplotlib 색상표


```python
# 두 개의 막대그래프 그리기

labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(men_means))
width = 0.35 # 막대 그래프 넓이 : 남자/여자 그래프의 폭을 지정했으므로 남/여 그래프의 중심값을 찾아내어 x 좌표로 사용해야한다. 어떻게 계산할 것인가?
# 남/여 그래프의 중심값 찐 중심값 -/+ (0.35/2)
plt.bar(x-width/2, men_means, width=width, color='dodgerblue', label='Men')
plt.bar(x+width/2, women_means, width=width, color='violet', label='Women')

plt.xlabel('City')
plt.ylabel('Score')
plt.xticks(x, labels) # x축 눈금의 값을 숫자가 아닌 문자로 표기하는 법
plt.title('Score by group and gender')
plt.legend()

plt.show()


# y = np.arange(len(women_means))
```

# 산점도 (Scatter Plot) 그리기
- 산점도 : 직교 좌표계를 이용해 두 개 변수 간의 상관관계를 나타내는 플롯

```python
x1 = [2,3,4]
y1 = [5,5,5]

x2 = [1,2,3,4,5]
y2 = [2,3,2,3,4]

y3 = [6,8,7,8,7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker='v', color='g')
plt.scatter(x2, y3, marker='^', color='r')

plt.title('Scatter Plot Example')
plt.show()
```

```python
np.random.seed(0)

for color in ['blue', 'orange', 'green']:
    n = 50
    x, y = np.random.rand(2, n)
    # [0,1) 0이상 1미만의 난수를 2행 50열로 생성
    # 첫 번째 행은 x값에, 두 번쨰 행은 y값에 할당
    scale = 190.0 * np.random.rand(n)+10
    # [10, 200) 190*0+10이상 190*1+10미만의 난수를 50개 생성
    plt.scatter(x, y, c=color, s=scale, label=color, alpha=0.3, edgecolors='none')
    # s=scale : 50개의 점이 각각 다른 형태와 크기로 찍힌다

plt.legend()
plt.grid(True) # 그래프에 격자 무늬 삽입

plt.show()
# 총 150개의 점이 그려진 것
```

# 히스토그램 (histogram) 그리기
- hist() : 히스토그램 생성
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 속하는 데이터의 빈도수를 표로 나타내는 것을 도수분포표라고 한다.
- 히스토그램은 도수분포표의 구간별 빈도수를 막대 그래프로 나타낸 것이다.
- 즉, 데이터의 구간을 설정하고 그 구간 안에 속해있는 데이터의 개수를 막대 그래프로 표현

```python
y = np.random.randn(1000) # 평균이 0이고, 표준편차가 1인 표준정규분포 난수를 1000개 생성

x = np.arange(len(y))


# 막대 그래프로 그렸을 때와 히스토그램으로 그렸을 때의 차이
plt.bar(x, y)
plt.title('Bar Chart')
plt.show() # 1000개의 각 값을 막대그래프로 그린 것

plt.hist(y, bins=50) # bins : 표현하고자 하는 데이터를 50개의 구간으로 나누기
plt.title('Histogram')
plt.show() # 각 구간별로 1000개 값의 빈도수을 히스토그램으로 그린 것



# 누적도수분포표에 대한 히스토그램을 그리기
plt.hist(y, cumulative=True, bins=20)
plt.title('Cumulative Histogram')
plt.show() # 각 구간별로 1000개 값의 누적 빈도수을 히스토그램으로 그린 것
```

# 원 그래프 (pie chart) 그리기
- pie() : 원 그래프 생성

```python
labels = ['Python', 'Java', 'C']

# autopct : 원 그래프의 조각마다 wedge 위에 표시할 데이터의 비율 레이블 (DEFAULT : None) # 몇 퍼센트를 차지하는 조각이냐?
plt.pie([0.3, 0.35, 0.4], labels=labels, autopct='%.1f%%')
# .1f : 소수점 한자리까지 비율을 표시
# %% : 퍼센트 표기시 앞에 % 하나 더 붙여주기

plt.title('Popular Language Ranking')
plt.show()
# 3시를 0도(기준 시작점, 시작 앵글)로 하여 반시계 방향으로 생성

# autopct : 총 합이 100이 될 필요가 없으며, 전체 총합에 대한 비율을 알아서 계산하여 보여줌
```

```python
section = [0.3, 0.35, 0.4]
labels = ['Python', 'Java', 'C']
color = ['coral', 'dodgerblue', 'yellowgreen']


plt.pie(section, # 조각의 크기를 설정하는 기본값
       explode=(0.1, 0, 0), # 조각을 중심으로부터 얼마만큼 떨어지게 할 것인지
        # python은 중심으로부터 반지름의 10%만큼 벗어남. 나머지는 그대로
        labels=labels, # 조각의 이름 표시
        colors=color, # 조각의 색상 결정
        autopct='%.2f%%', # 비율 표시 포맷 설정
        shadow=True, # 그림자 표시 및 입체감 생성
        startangle=90, # 원형의 시작 부분을 각도로 표시 (DEFAULT 0도, 3시 방향)
        counterclock=False, # 반시계방향으로 순서대로 표시할 지 결정 (DEFAULT : T 반시계 / F 시계)
        rotatelabels=True #조각의 이름을 회전
       )

plt.title('Pie Chart')
plt.show()
```

# 상자 수염 (boxplot) 그리기
- boxplot : 많은 데이터에서 데이터 집합의 범위와 중앙값 등 통계 수치 값 확인 목적
- 이상치(outlier)의 유무 확인이 용이
- IQR (Inter Quartile Range, 사분위 범위) = Q3 - Q1
    - 최소값 : 1사분위수 - 1.5 * IQR
    - 1사분위(Q1) : 중앙값 기준으로 하위 50%의 중앙값 / 전체 데이터 중 하위 25%에 해당하는 값
    - 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미 / 전체 데이터 중 정 가운데 순위에 해당하는 값
    - 3사분위(Q3) : 중앙값 기준으로 상위 50%의 중앙값 / 전체 데이터 중 상위 25%에 해당하는 값
    - 최대값 : 3사분위수 + 1.5 * IQR

```python
np.random.seed(10)

# random.normal : 평균과 표준편차 지정 가능
data_a = np.random.normal(0, 2.0, 1000) # 평균:0, 표준편차:2인 1000개의 난수
data_b = np.random.normal(-3.0, 1.5, 500) #평균:-3, 표준편차:1.5인 500개 난수
data_c = np.random.normal(1.2, 1.5, 1500) #평균:1.2, 표준편차:1.5인 1500개 난수

plt.boxplot([data_a, data_b, data_c])
plt.ylim(-10.0, 10.0)
plt.xlabel('Data Type')
plt.ylabel('Value')

plt.show()
# 상자의 길이 : IQR
# 주황색 : Q2 / 중앙값 median
# 상자 윗면 : Q3
# 상자 아래면 : Q1
# 수염 길이 : 1.5 * IQR
# 수염 위 : 최대값 Max
# 수염 아래 : 최소값 Min
# 동그라미 : 이상치 outlier
```

- Matplotlib 라이브러리 : numpy 다차원 배열만 시각화 가능함
- 씨본 라이브러리 : 데이터프레임 바로 전달해도 시각화 가능함
