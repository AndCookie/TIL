# 딥러닝 (Deep Learning)

## 퍼셉트론 (Perceptron)
- 초기의 인공신경망 (실제 뇌를 구성하는 신경 세포 뉴런의 동작과 유사)
- many to one : 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘
- 
        - x : 입력값
        - w : 가중치 weight
        - y : 출력값
        - 원 : 인공 뉴런
        - 선 : 신호 전달하는 축삭돌기
    - 입력값마다 가중치가 존재하며, 가중치의 값이 클수록 해당 입력 값이 중요하다는 의미

### 활성화 함수 (Activation Function)
- 활성화 함수 : 뉴런에서 출력값을 변경시키는 함수
- 1. 계단 함수 (Step Function)
    - 
    - 각 입력값이 가중치와 곱해져서 인공 뉴런에 보내지고, 각 입력값과 그에 해당되는 가중치의 곱의 전체 합이 임계치(θ 세타, threshold)를 넘으면 종착지에 있는 인공 뉴런은 출력 신호로서 1을 출력하고, 그렇지 않을 경우에는 0을 출력
    - 
- 2. 시그모이드 함수
- 3. 소프트맥수 함수

### 단층 퍼셉트론 (SLP, Single-Layer Perceptron)
- 두 단계(층 layer)로 구성된 퍼셉트론
    - 입력층(input layer) : 값을 보내는 단계
    - 출력층(output layer) : 값을 받아서 출력하는 단계
- 기능 :
    - 1. AND 게이트 : 구현 가능
        - 두 개의 입력값 x1, x2이 각각 0 또는 1의 값을 가질 수 있으며
        - 모두 1인 경우에만 출력값 y가 1인 구조
    - 2. NAND 게이트 : 구현 가능
        - 두 개의 입력값 x1, x2이 각각 0 또는 1의 값을 가질 수 있으며
        - 모두 1인 경우에만 출력값 y가 0, 나머지 경우에는 출력값 y가 1인 구조
    - 3. OR 게이트 : 구현 가능
        - 두 개의 입력값 x1, x2이 각각 0 또는 1의 값을 가질 수 있으며
        - 모두 0인 경우에 출력값 y가 0, 나머지 경우에는 출력값 y가 1인 구조
    - 4. XOR 게이트 : 구현 불가능
        - 두 개의 입력값 x1, x2이 각각 0 또는 1의 값을 가질 수 있으며
        - 입력값 두 개가 서로 다른 값이면 출력값 y가 1, 같은 값이면 출력값 y가 0인 구조
        - 대안 :다층 퍼셉트론을 사용하여 여러 선으로 분류하는 방식으로 보완하기

### 다층 퍼셉트론 (MLP, MultiLayer Perceptron)
- 여러 단계(층 layer)로 구성되며 은닉층이 1개 이상인 퍼셉트론
   - 입력층(input layer) : 값을 보내는 단계
   - 은닉층(hidden layer) :
   - 출력층(output layer) : 값을 받아서 출력하는 단계
- 기능 :
   - AND, NAND, OR, XOR 모두 구현 가능

- 딥러닝 (Deep Learning)
    - 심층 신경망을 학습시키는 것
- 심층 신경망 (DNN, Deep Neural Network)
    - 은닉층이 2개 이상인 신경망

## 인공 신경망 (Artificial Neural Network)

### 신경망의 종류 (FFNN, RNN)
- 1. 순방향 신경망 (FFNN, Feed-Forward Neural Network)
    - 입력층 -> 출력층 방향으로 연산이 전개되는 신경망
    - 다층 퍼셉트론(MLP)과 동일)
- 2. 순환 신경망 (RNN, Recurrent Neural Network)
    - 입력층 <-> 출력층으로 연산이 전개되는 신경망
    - 은닉층의 출력값을 출력층으로도 값을 보내지만, 동시에 은닉층의 출력값이 다시 은닉층의 입력으로 사용됨
    - FFNN에 속하지 않는 신경망

### 전결합층 (FC, Fully-connected layer, Dense layer)
- = 완전연결층 = 밀집층 (Dense Layer)
- 어떤 층의 모든 뉴런이 이전 층의 모든 뉴런과 연결돼 있는 층
- 다층 퍼셉트론의 모든 은닉층과 출력층은 전결합층임

### 활성화 함수 (Activation Function)
- 은닉층 : ReLU(또는 ReLU 변형) 함수들을 사용하는 것이 일반적
- 출력층 : 시그모이드 함수(이진분류-로지스틱회귀), 소프트맥스 함수(다지분류-소프트맥스회귀)

< 특징 >
-  인공 신경망에서의 활성화 함수는 "비선형 함수 (Nonlinear function)"
    - 선형 함수 : 출력이 입력의 상수배만큼 변하는 함수
        - ex) f(x) = wx+b 의 경우, 상수는 w와 b이며 직선으로 시각화됨
    - 비선형 함수 : 직선 1개로는 그릴 수 없는 함수
       - 인공 신경망에서의 활성화 함수 (계단 함수 포함)
       - 인공 신경망의 능력을 높이기 위해서는 은닉층을 계속해서 추가해야 하는데, 만약 활성화 함수로 선형 함수를 사용하게 되면 은닉층을 쌓을 수가 없기 때문
- 인공 신경망은 선형층과 비선형층으로 구성됨
    - 1. 선형층 (linear layer)
        - = 투사층 (projection layer)
        - 활성화 함수가 존재하지 않는 선형 함수층
        - ex) 임베딩 층 (embedding layer)
    - 2. 비선형층 (nonlinear layer)
        - = 은닉층
        - 활성화 함수가 존재하는 비선형 함수층

< 종류 >
1. 계단 함수 (Step function)
    - 거의 사용되지 않지만 퍼셉트론을 통해 인공 신경망을 처음 배울 때 접하게 되는 활성화 함수
    
2. 시그모이드 함수 (Sigmoid function)
    - 입력값을 0과 1사이의 값으로 변환
    - 가중치와 편향을 업데이트하는 역전파 과정에서 기울기 소실 문제 발생 (Vanishing Gradient)
        - 손실(오차 loss)을 미분하여 기울기(gradient)를 계산하는 단계에서
        - 시그모이드 함수를 사용하는 은닉층의 개수가 다수가 될 경우
        - 0에 가까운 기울기 값이 누적해서 곱해지게 되면서, 출력층과 가까운 은닉층에서는 기울기(미분값)가 잘 전파되지만, 앞단으로 갈수록 기울기가 제대로 전달되지 않음
        - 매개변수 w 업데이트 불가
        - 즉, 학습 불가
    - 주로 이진 분류를 위해 출력층에서 사용됨
    
    
3. 하이퍼볼릭탄젠트 함수 (Hyperbolic tangent function)
    - 입력값을 -1과 1사이의 값으로 변환
    - -1과 1에 가까운 출력값을 출력시, 시그모이드 함수처럼 기울기 소실 문제가 발생하지만 그 증상이 적음
        - 시그모이드 함수와는 달리 0을 중심으로 하고있으며, 하이퍼볼릭탄젠트 함수를 미분했을 때의 최대값은 1로 시그모이드 함수의 최대값인 0.25보다는 크기 때문에 미분 결과값이 시그모이드 함수보다는 전반적으로 크다보니 기울기 소실 증상이 적은 편이라 은닉층에서 시그모이드 함수보다는 선호됨
    

4. 렐루 함수 (ReLU)
    - f(x) = max (0, x)
    - 음수를 입력하면 0을 출력하고, 양수를 입력하면 입력값을 그대로 반환
        - 출력값이 특정 양수값에 수렴하지 않으며
        - 0이상의 입력값의 경우에는 미분값이 항상 1
    - 인공 신경망의 은닉층에서 가장 인기있는 함수
    - 깊은 신경망의 은닉층에서 시그모이드 함수보다 훨씬 더 잘 작동
    - 장점 : 빠른 연산 속도
        - 시그모이드 함수와 하이퍼볼릭탄젠트 함수와 같이 어떤 연산이 필요한 것이 아니라 단순 임계값이므로
    - 한계 : 죽은 렐루(dying ReLU)
        - 입력값이 음수면 기울기(미분값)도 0이 되며,이 뉴런은 다시 회생하는 것이 매우 어려움
    

5. 리키 렐루 (Leaky ReLU)
    - f(x) = max (ax, x)
    - 음수를 입력하면 0이 아닌 0.001과 같은 매우 작은 수를 출력하고, 양수를 입력하면 입력값을 그대로 반환
    - 죽은 렐루를 보완하기 위한 ReLU의 변형 함수 (leaky : 새는)
        - a : Leaky('새는') 정도를 결정하는 하이퍼파라미터 / 입력값의 음수일 때의 기울기로 일반적으로 0.01의 값을 가짐
        - 입력값이 음수라도 기울기가 0이 되지 않으면 ReLU는 죽지 않음
    
        
6. 소프트맥스 함수 (Softmax Function)
    - 세 가지 이상의 (상호 배타적인) 선택지 중 하나를 고르는 다중 클래스 분류(MultiClass Classification) 문제에 주로 사용
    

### 행렬곱으로 이해하는 신경망
- 1. 순전파 (Foward Propagation)
    - 인공 신경망에서 입력층 -> 은닉층 -> 출력층 방향으로 예측값 연산을 진행하는 과정
    - 입력값은 입력층과 은닉층을 지나면서 각 층에서의 가중치와 함께 연산되며, 출력층에서 모든 연산을 마친 예측값이 나옴
- 2. 신경망의 순전파는 결과적으로 행렬의 곱셈으로 이해 가능
    - 
- 3. 배치 연산 : 인공 신경망이 다수의 샘플을 동시에 처리하는 것
- 4. 행렬 크기 추정 : 배치 크기=1로 가정하여 층마다 생기는 가중치와 편향 행렬의 크기를 추정 (학습 가능한 매개변수인 가중치 w와 편향 b의 개수를 추정하는 방법)

```python
# 인공 신경망을 케라스로 구현하기
# 입력의 차원이 3, 출력의 차원이 2인 위 인공 신경망 구현하기

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# 3개의 입력과 2개의 출력
model.add(Dense(2, input_dim=3, activation='softmax'))
# 활성화 함수 : 소프트맥수 함수 (이진분류와 다지분류 모두 수행 가능)
# 소프트맥스 회귀는 출력 벡터의 차원을 2로 두면 이진 분류를 수행하는 모델이 됨
# 활성화 함수는 행렬의 크기에 영향을 주지 않음
```

```python
model.summary()

# keras의 summary : 해당 모델에 존재하는 모든 매개변수(가중치 w와 편향 b의 개수)를 확인

# 매개변수의 개수 : 8 = 3 * 2 + 2 = 가중치 w 6개 + 편향 b 2개
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# 4개의 입력과 8개의 출력
model.add(Dense(8, input_dim=4, activation='relu'))

# 이어서 8개의 출력
model.add(Dense(8, activation='relu'))

# 이어서 3개의 출력
model.add(Dense(3, activation='softmax'))

# 입력층 : 4개의 입력과 8개의 출력
# 은닉층1 : 8개의 입력과 8개의 출력
# 은닉층2 : 8개의 입력과 3개의 출력
# 출력층 : 3개의 입력과 3개의 출력
```

## 딥러닝의 학습 방법
1. 순전파 (forward propagation)
    - 입력에 대해서 입력층 -> 출력층 방향으로 연산
2. 역전파 (back propagation)
    - 출력층 -> 입력층 방향으로 가중치와 편향을 업데이트
    - 1. 오차 계산
        - 순전파 연산을 통해 나온 예측값과 실제값의 오차를 손실 함수(loss function)을 통해 계산
    - 2. 기울기 계산
        - 손실(오차 loss)을 미분을 통해서 기울기(gradient) 계산
    - 3. 경사하강법으로 가중치와 편향 업데이트
        - 옵티마이저를 지정하여 특정 경사하강법으로 학습률(이동거리)에 따라 가중치를 수정 및 학습시키기
4. 순전파
    - 업데이트된 가중치에 대해 다시 순전파를 진행하여 오차 감소 여부 확인
5. 역전파
    - 오차를 최소화하는 가중치를 찾을 때까지 순전파와 역전파를 반복

- 중요 포인트
 : 업데이트가 시작되는 시점은 실제값과 예측값을 확인하는 시점

## 순전파

### 손실 함수 (Loss Function)
< 정의 >
- 실제값과 예측값의 차이를 수치화해주는 함수

< 특징 > 
- 오차가 클수록 손실 함수의 값은 크고, 오차가 작을수록 손실 함수의 값은 작아짐

< 목적 >
- 딥러닝의 학습 과정 = 손실 함수의 값을 최소화하는 두 개의 매개변수인 가중치 w와 편향 b의 값을 찾는 것

< 종류 >
- 딥러닝 자연어 처리는 대부분 분류 문제이므로 평균 제곱 오차보다는 크로스 엔트로피 함수들을 주로 사용
- 회귀 :
    - 1. 평균 제곱 오차 MSE (Mean Squared Error) 
        - 연속형 변수 예측시 사용
- 분류 : 크로스 엔트로피
    - 2. 이진 크로스 엔트로피 (Binary Cross-Entropy)
        - = 이항 교차 엔트로피
        - 출력층에서 시그모이드 함수를 사용하는 이진 분류 (Binary Classification)의 경우 binary_crossentropy를 사용
    - 3. 카테고리칼 크로스 엔트로피 (Categorical Cross-Entropy)
        - = 범주형 교차 엔트로피
        - 출력층에서 소프트맥스 함수를 사용하는 다중 클래스 분류(Multi-Class Classification)일 경우 categorical_crossentropy를 사용
        - sparse_categorical_crossentropy : 만약 레이블(실제값 y)에 대해서 원-핫 인코딩 과정을 생략하고, 정수값을 가진 레이블에 대해서 다중 클래스 분류를 수행하고 싶을 때 사용

```python
# MSE (Mean Squared Error) 
model.compile(optimizer='adam', loss='mse', metrics=['mse'])

# model.compile(optimizer='adam', loss=tf.keras.losses.MeanSquaredError(), metrics=['mse'])
```

```python
# 이진 크로스 엔트로피 (Binary Cross-Entropy)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])

# model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer='adam', metrics=['acc'])
```

```python
# 카테고리칼 크로스 엔트로피(Categorical Cross-Entropy)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])

# model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer='adam', metrics=['acc'])
```

```python
# sparse_categorical_crossentropy
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])

#  model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(), optimizer='adam', metrics=['acc'])
```

### 옵티마이저 (Optimizer)
< 정의 >
- 손실 함수의 값을 줄여나가는 학습 방법
- 경사하강법 알고리즘

< 용어 >
- 1. 배치 (Batch)
    - 가중치 등의 매개 변수의 값을 조정(업데이트)하기 위해 사용하는 데이터의 수량 단위
    - 전체 데이터 또는 정해준 양의 데이터로 매개 변수의 값 조정 가능
    - 전체 데이터 크기 / 배치 크기(Batch Size) = 배치 수(Iteration) 

- 2. 에포크 (Epoch)
    - 전체 데이터에 대한 훈련 횟수
    - 인공 신경망에서 전체 데이터에 대해서 순전파와 역전파가 끝난 상태
    - 에포크 횟수가 지나치게 많거나 적으면 과적합과 과소적합 발생 가능
   
- 3. 이터레이션 (Iteration) 또는 스텝(Step)
    - 한 번의 에포크를 끝내기 위해서 필요한 배치의 수
    - 한 번의 에포크 내에서 이루어지는 매개변수의 업데이트 횟수 (한 번의 에포크 당 매개변수 업데이트가 n번 이루어진다는 의미)
        - 배치 크기가 1인 확률적 경사하강법 : 모든 이터레이션마다 하나의 데이터를 선택하여 경사 하강법을 수행



< 경사하강법의 종류 >
- 1. 배치 경사 하강법 (Batch Gradient Descent)
    - 오차(loss) 계산시 전체 데이터를 고려
    - 한 번의 에포크에 모든 매개변수 업데이트를 단 한 번 수행
    - 장점 : 전체 데이터를 고려해서 학습하므로 정확도가 높음
    - 단점 : 한 번의 매개 변수 업데이트에 시간이 오래 걸리며, 메모리를 크게 요구함
    
- 2. 배치 크기가 1인 확률적 경사 하강법 (SGD, Stochastic Gradient Descent)
     - 매개변수 값을 조정 시 전체 데이터가 아니라 랜덤으로 선택한 하나의 데이터에 대해서만 계산하는 방법
     - 더 적은 데이터를 사용하므로 더 빠르게 계산 가능
     - 장점 : 하나의 데이터에 대해서만 메모리에 저장하면 되므로 자원이 적은 컴퓨터에서도 쉽게 사용 가능
     - 단점 : 매개변수의 변경폭이 불안정하고, 때로는 배치 경사 하강법보다 정확도가 낮을 수 있음

- 3. 미니 배치 경사 하강법 (Mini-Batch Gradient Descent)
    - 배치 크기를 지정하여 해당 데이터 개수만큼에 대해서 계산하여 매개 변수의 값을 조정하는 경사 하강법
    - 가장 많이 사용되는 경사 하강법으로 일반적으로 배치 크기는 2의 n제곱에 해당하는 숫자로 선택
    - model.fit()에서 배치 크기를 별도로 지정해주지 않을 경우에 기본값은 2의 5제곱에 해당하는 숫자인 32로 설정
    - 장점 : 전체 데이터를 계산하는 것보다 빠르며, SGD보다 안정적

< 옵티마이저 >
- 1. 모멘텀 (Momentum)
    - 경사 하강법에서 계산된 접선의 기울기에 한 시점 전의 접선의 기울기값을 일정한 비율만큼 반영
    - 로컬 미니멈에 빠지더라도 관성의 힘으로 탈출하여, 글로벌 미니멈 내지는 더 낮은 로컬 미니멈 찾는 효과
 
- 2. 아다그라드(Adagrad)
    - 각 매개변수에 서로 다른 학습률을 적용
        - 변화가 많은 매개변수는 학습률을 작게 설정
        - 변화가 적은 매개변수는 학습률을 높게 설정
    - 매개변수들은 각자 의미하는 바가 다른데, 모든 매개변수에 동일한 학습률(learning rate)을 적용하는 것은 비효율적이기 때문
    - 단점 : 계속 학습 진행시, 나중에 가서는 학습률이 지나치게 떨어짐

- 3. 알엠에스프롭 (RMSprop)
    - 아다그라드의 단점을 보완
    
- 4. 아담(Adam)
    - 알엠에스프롭과 모멘텀 두 가지를 합친 듯한 방법
    - 방향과 학습률 두 가지를 모두 잡기 위한 방법

## 과적합 (Overfitting)
< 정의 >
- 불필요하고 과도한 학습으로 인해 훈련 데이터에 대한 정확도는 높지만 검증 데이터나 테스트 데이터에 대한 정확도는 낮은 현상
- 모델 성능 저하 원인

< 대안 >
- 1. 데이터 양 늘리기
    - 모델은 데이터의 양이 적을 경우, 해당 데이터의 특정 패턴이나 노이즈까지 쉽게 암기하기 되므로 과적합 발생 가능
    - 데이터의 양을 늘릴 수록 모델은 데이터의 일반적인 패턴을 학습하므로 과적합 방지 가능
        - 데이터 증식 또는 증강 (Data Augmentation)
            - 데이터의 양이 적을 경우, 의도적으로 기존의 데이터를 조금씩 변형하고 추가하여 데이터의 양을 늘리는 방법
            - 텍스트 데이터 : 역번역(Back Translation) : 번역 후 재번역을 통해 새로운 데이터를 만들어냄
            - 이미지 데이터 : 이미지를 돌리거나 노이즈를 추가하고, 일부분을 수정하여 데이터 증식
            
- 2. 모델의 복잡도 줄이기
    - 은닉층(hidden layer)과 매개변수의 개수 줄이기
    - 모델의 수용력(capacity) : 모델 내 매개변수의 개수
    
- 3. 가중치 규제(Regularization) 적용하기
    - 모델 간단하게 만드는 방법 (매개변수가 많은 복잡한 모델이 매개변수가 적은 간단한 모델보다 과적합될 가능성이 높으므로)
        - 람다 λ : 규제의 강도를 정하는 하이퍼파라미터
            - 람다가 클수록 모델이 훈련 데이터에 대해서 적합한 매개 변수를 찾는 것보다 규제를 위해 추가된 항들을 작게 유지하는 것을 우선한다는 의미
        - L1 규제 (L1 노름)
            - 가중치 w들의 절대값 합계를 비용 함수에 추가
            - 영향력이 적은 가중치들은 0이 되어 사라짐
        - L2 규제 (L2 노름)
            - 모든 가중치 w들의 제곱합을 비용 함수에 추가
            - 영향력이 적은 가중치들은 0에 가까워지지만 완전히 0이 되지는 않음
            - 가중치 감쇠 (weight decay)
            
- 4. 드롭아웃 (Dropout)
    - 학습 과정에서 신경망의 일부를 사용하지 않는 방법
    - 신경망 학습 시에만 사용하고, 예측 시에는 사용하지 않는 것이 일반적
    - 학습 시에 인공 신경망이 특정 뉴런 또는 특정 조합에 너무 의존적이게 되는 것을 방지
    - 매번 랜덤 선택으로 뉴런들을 사용하지 않으므로 서로 다른 신경망들을 앙상블하여 사용하는 것 같은 효과를 내어 과적합을 방지

```python
# 모델에 드롭아웃 추가하기

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Dense

max_words = 10000
num_classes = 46

model = Sequential()
model.add(Dense(256, input_shape=(max_words,), activation='relu'))
model.add(Dropout(0.5)) # 드롭아웃 추가. 비율은 50%
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5)) # 드롭아웃 추가. 비율은 50%
model.add(Dense(num_classes, activation='softmax'))
```

## 기울기 소실과 폭주 (Gradient Vanishing and Exploding)

< 정의 >
1. 기울기 소실
    - 역전파 과정에서 입력층으로 갈수록 기울기(Gradient)가 점차적으로 작아지는 현상
    - 기울기 소실에 의해 입력층에 가까운 층들에서 가중치들이 업데이트가 제대로 되지 않으면 결국 최적의 모델을 찾을 수 없게 됨

2. 기울기 폭주
    - 기울기(Gradient)가 점차 커지더니 가중치들이 비정상적으로 큰 값이 되면서 결국 발산
    - 순환 신경망(Recurrent Neural Network, RNN)에서 쉽게 발생

< 방지법 >
1. ReLU 활용하기
- 은닉층의 활성화 함수로 ReLu나 Leaky ReLu 사용하기
- 시그모이드나 하이퍼볼릭탄젠트 함수는 입력의 절대값이 클 경우에 출력값이 0 또는 1에 수렴하면서 기울기가 0에 가까워지다보니, 역전파 과정에서 전파시킬 기울기가 점차 사라져서 입력층 방향으로 갈수록 제대로 역전파가 되지 않는 기울기 소실 문제가 발생하므로 은닉층에서 사용하지 말기
- Leaky ReLU : 모든 입력값에 대해서 기울기가 0에 수렴하지 않아 죽은 ReLU 문제를 해결

2. 그래디언트 클리핑 (Gradient Clipping)
- 임계치만큼 기울기 값을 잘라 크기를 감소시키는 것
- 기울기 폭주를 막기 위해 임계값을 넘지 않는 선에서 자르기
- RNN에서 유용 : 역전파 과정에서 시점을 역행하면서 기울기를 구하는데 이때 기울기가 너무 커질 수 있기 때문

3. 가중치 초기화 (Weight Initialization)
- 활성화 함수로 시그모이드나 하이퍼볼릭탄젠트 함수 사용시 : 세이비어 초기화 방법이 효율적
- 활성화 함수로 ReLu나 Leaky ReLu 함수 사용시 : He 초기화 방법이 효율적

- 1. 세이비어 초기화 (Xavier Initialization)
    - = 글로럿 초기화(Glorot Initialization)
    - 여러 층의 기울기 분산 사이에 균형을 맞춰서 특정 층이 너무 주목을 받거나 다른 층이 뒤쳐지는 것을 방지
    - 시그모이드 함수나 하이퍼볼릭 탄젠트 함수와 같은 S자 형태인 활성화 함수와 함께 사용할 경우에는 좋은 성능을 보임
    - 그러나 ReLU와 함께 사용할 경우에는 성능이 좋지 않음 -> He 초기화(He initialization)
        - n in : 이전 층의 뉴런 개수
        - n out : 다음 층의 뉴런 개수
    - 1. 균등 분포 (Uniform Distribution)
        - 균등 분포로 가중치를 초기화할 경우 다음과 같은 균등 분포 범위를 사용
        - 
    - 2. 정규 분포 (Normal Distribution)
        - 정규 분포로 가중치를 초기화할 경우, 평균이 0이고 표준 편차 σ가 다음을 만족하도록 함
        - 
        
- 2. He 초기화 (He Initialization)
    - 세이비어 초기화와 다르게 다음 층의 뉴런의 수를 반영하지 않음
    - ReLu 함수나 Leaky ReLu 함수시 효율적
        - n in : 이전 층의 뉴런 개수
    - 1. 균등 분포 (Uniform Distribution)
        - 균등 분포로 가중치를 초기화할 경우 다음과 같은 균등 분포 범위를 사용
        - 
    - 2. 정규 분포 (Normal Distribution)
        - 정규 분포로 가중치를 초기화할 경우, 표준 편차 σ가 다음을 만족하도록 함
        - 
        
4. 배치 정규화 (Batch Normalization)
- 인공 신경망의 각 층에 들어가는 입력을 평균과 분산으로 정규화하여 효율적 학습 가능
- 활성화 함수 ReLU + He 초기화 방법 조합으로 어느 정도 기울기 소실과 폭주를 완화시킬 수 있지만, 이 두 방법을 사용하더라도 훈련 중에 언제든 다시 발생 가능
- 1. 내부 공변량 변화 (Internal Covariate Shift)
    - 학습 과정에서 층 별로 입력 데이터 분포가 달라지는 현상
    - 이전 층들의 학습에 의해 이전 층의 가중치 값이 바뀌게 되면, 현재 층에 전달되는 입력 데이터의 분포가 현재 층이 학습했던 시점의 분포와 차이가 발생
    - 공변량 변화 : 훈련 데이터의 분포와 테스트 데이터의 분포가 다른 경우
    - 내부 공변량 변화 : 신경망 층 사이에서 발생하는 입력 데이터의 분포 변화
- 2. 배치 정규화 (BN, Batch Normalization)
    - 한 번에 들어오는 배치 단위로 정규화하는 것
    - 각 층에서 활성화 함수를 통과하기 전에 수행
        - 1. 입력에 대해 평균을 0으로 만들고 정규화
        - 2. 정규화된 데이터에 대해 스케일과 시프트를 수행
            - 매개변수 γ로 스케일
            - 매개변수 β로 시프트
        - 3. 다음 레이어에 일정한 범위의 값들만 전달
    - 학습시 배치 단위의 평균과 분산들을 차례대로 받아 이동 평균과 이동 분산을 저장
    - 테스트시 해당 배치의 평균과 분산을 구하지 않고, 학습 단계에서 저장해두었던 평균과 분산으로 정규화
    - 시그모이드 함수나 하이퍼볼릭탄젠트 함수를 사용하더라도 기울기 소실 문제가 크게 개선됨
    - 가중치 초기화에 훨씬 덜 민감해짐
    - 훨씬 큰 학습률을 사용할 수 있어 학습 속도 개선
    - 미니 배치마다 평균과 표준편차를 계산하여 사용하므로 훈련 데이터에 일종의 잡음 주입의 부수 효과로 과적합을 방지하는 효과, 즉 마치 드롭아웃과 비슷한 효과
    - 모델을 복잡하게 하며, 추가 계산을 하는 것이므로 테스트 데이터에 대한 예측 시에 실행 시간이 느려짐
    - 수식
        - input : 미니 배치 B = {x(1), x(2), .., x(m)}, 
        - output : y(i) = BN γ,β (x(i))
            - 미니 배치 : 동일한 특성(feature) 개수들을 가진 다수의 샘플들
            - m : 미니 배치에 있는 샘플의 수
            - μB : 미니 배치 B의 평균
            - σB : 미니 배치 B의 표준편차
            - x^(i) : 평균이 0이고 정규화된 입력 데이터
            - ε : σ의 제곱이 0일 때, 분모가 0이 되는 것을 방지하기 위한 작은 양수로서 보편적으로 10의 -5승
            - γ : 정규화된 데이터에 대한 스케일 매개변수로서 학습 대상임
            - β : 정규화된 데이터에 대한 시프트 매개변수로서 학습 대상임
            - y(i) : 스케일과 시프트를 통해 조정한 BN의 최종 결과
        - 미니 배치의 평균 계산

    - 한계
        - 1. 미니 배치 크기에 의존적

## 워드 임베딩 (Word Embedding)
텍스트 내의 단어들을 밀집 벡터(dense vector)로 만드는 것


- 1. 원-핫 인코딩
    - : 단어를 원-핫 벡터로 만드는 것
    - 원-핫 벡터
        - 대부분의 원소의 값이 0인 희소 벡터
        - 원소 : 0 또는 1
        - 고차원 : 단어 집합의 크기만큼 벡터의 차원을 가짐
        - 단어 벡터 간의 유의미한 유사도를 구할 수 없음

- 2. 워드 임베딩
   - : 단어를 밀집 벡터(dense vector)로 만드는 것
    - 임베딩 벡터
        - 모든 원소의 값이 실수인 밀집 벡터
        - 원소 : 실수
        - 저차원
        - 훈련 데이터로부터 학습 : 초기에는 랜덤값을 가지지만, 인공 신경망의 가중치가 학습되는 방법과 같은 방식으로 값이 학습되며 변경

## 컴파일(Compile)과 훈련(Training)
- 임베딩층, 은닉층, 출력층을 추가하여 모델을 설계한 후에, 마지막으로 컴파일

### compile()
- 기능 : 모델을 기계가 이해할 수 있도록 컴파일
    - 손실 함수와 옵티마이저, 메트릭 함수를 선택
- compile (optimizer='옵티마이저', loss='손실함수', metrics=['모니터링 지표'])
    - optimizer : 훈련 과정을 설정하는 옵티마이저를 설정
    - loss : 훈련 과정에서 사용할 손실 함수(loss function)를 설정
    - metrics : 훈련을 모니터링하기 위한 지표를 선택
- 손실함수와 활성화함수 조합
    - 회귀 문제
        - 손실함수 : mse
        - 활성화함수 : X
        - ex) 선형 회귀
    - 이진 분류
        - 손실함수 : binary_crossentropy
        - 활성화함수 : 시그모이드
        - ex) 로이터 뉴스 분류하기
    - 다지 분류
        - 손실함수 : categorical_crossentropy
        - 활성화함수 : 소프트맥스
        - ex) 양방향 LSTM를 이용한 품사 태깅
    - 다지 분류
        - 손실함수 : sparse_categorical_crossentropy
            - 레이블을 원-핫 인코딩하지 않고 정수 인코딩된 상태에서 수행 가능
        - 활성화함수 : 소프트맥스
        - ex) IMDB 리뷰 감성 분류하기

```python
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense
from tensorflow.keras.models import Sequential

vocab_size = 10000
embedding_dim = 32
hidden_units = 32

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

### fit()
- 모델 학습
- 학습 = 훈련 = 적합(fitting) = 모델이 오차로부터 매개 변수를 업데이트 시키는 과정
- fit(훈련데이터, 레이블데이터, epochs=총 훈련 횟수, batch_size=배치크기, verbose=학습 중 출력 문구, validation_data(x_val, y_val), validation_split=비율)
    - batch_size : 기본값은 32 / 미니 배치 경사 하강법을 사용하고 싶지 않을 경우에는 batch_size=None을 기재
    - verbose : 학습 중 출력되는 문구 설정
        - 0 : 아무것도 출력하지 않음
        - 1 : 훈련의 진행도를 보여주는 진행 막대 보여줌
        - 2 : 미니 배치마다 손실 정보 보여줌
    - validation_data(x_val, y_val)
        - 검증 데이터(validation data)를 사용하기 위한 용도
        - 일반적으로 검증 데이터를 사용하면 각 에포크마다 검증 데이터의 정확도나 오차를 함께 출력하는데, 이 정확도는 훈련이 잘 되고 있는지를 보여줄 뿐이며 실제로 모델이 검증 데이터를 학습하지는 않음
        - 검증 데이터의 오차(loss)가 낮아지다가 높아지기 시작하면 이는 과적합(overfitting)의 신호
    - validation_split
        - validation_data와 동일하게 검증 데이터를 사용하기 위한 용도
        - 검증 데이터를 지정하는 것이 아니라 훈련 데이터와 훈련 데이터의 레이블인 X_train과 y_train에서 일정 비율 분리하여 이를 검증 데이터로 사용

# keras의 Subclassing API
- 장점
    - Functional API로도 구현할 수 없는 모델들조차 구현 가능
        - ex) 재귀 네트워크나 트리 RNN과 같이 딥 러닝 모델을 DAG(directed acyclic graph)로 취급하지 않는 경우 Functional API에서 구현 불가능하나 Subclassing API에서 구현 가능
    - 밑바닥부터 새로운 수준의 아키텍처를 구현해야 하는 실험적 연구를 하는 연구자들에게 적합
- 단점
    - 객체 지향 프로그래밍(Object-oriented programming)에 익숙해야 하므로 코드 사용이 가장 까다로움
 


## 선형 회귀

```python
import tensorflow as tf

class LinearRegression(tf.keras.Model):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.linear_layer = tf.keras.layers.Dense(1, input_dim=1, activation='linear')

    def call(self, x):
        y_pred = self.linear_layer(x)

        return y_pred
```

```python
model = LinearRegression()

X = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부하는 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 각 공부하는 시간에 맵핑되는 성적

sgd = tf.keras.optimizers.SGD(lr=0.01)
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])
model.fit(X, y, epochs=300)


# 클래스(class) 형태의 모델은 tf.keras.Model을 상속받음
# init()에서 모델의 구조와 동적을 정의하는 생성자를 정의 : 이는 파이썬에서 객체가 갖는 속성값을 초기화하는 역할로, 객체가 생성될 때 자동으호 호출됨
# super() 함수를 부르면 여기서 만든 클래스는 tf.keras.Model 클래스의 속성들을 가지고 초기화됨
# call() 함수는 모델이 데이터를 입력받아 예측값을 리턴하는 포워드(forward) 연산을 진행시키는 함수


# H(x) 식에 입력x로부터 예측된 y를 얻는 것을 forward 연산이라고 한다
```

### count
- 문서 단어 행렬(Document-Term Matrix, DTM)을 생성
- DTM에서의 인덱스는 앞서 확인한 word_index의 결과
- 주의 : 각 단어에 부여되는 인덱스는 1부터 시작하는 반면에 완성되는 행렬의 인덱스는 0부터 시작

```python
# 전처리

vocab_size = 10000 # 최대 단어 개수
num_classes = 20
```

```python
def prepare_data(train_data, test_data, mode): # 전처리 함수
    tokenizer = Tokenizer(num_words = vocab_size) # vocab_size 개수만큼의 단어만 사용한다.
    tokenizer.fit_on_texts(train_data)
    X_train = tokenizer.texts_to_matrix(train_data, mode=mode) # 샘플 수 × vocab_size 크기의 행렬 생성
    X_test = tokenizer.texts_to_matrix(test_data, mode=mode) # 샘플 수 × vocab_size 크기의 행렬 생성
    return X_train, X_test, tokenizer.index_word
```

## 다층 퍼셉트론(Multilayer Perceptron, MLP)을 사용하여 텍스트 분류하기
- 깊은 신경망(Deep Neural Network, DNN) : 2개의 은닉층을 포함하여 총 4개의 층으로 구성
    - vocab_size의 크기를 가진 입력층
    - 256개의 뉴런을 가진 첫번째 은닉층
    - 128개의 뉴런을 가진 두번째 은닉층
    - num_classes의 크기를 가진 출력층


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

def fit_and_evaluate(X_train, y_train, X_test, y_test):
    model = Sequential()
    model.add(Dense(256, input_shape=(vocab_size,), activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    # 과적합을 막기 위해서 두 번의 드롭아웃(Dropout)을 적용
    # 출력층의 활성화 함수로는 소프트맥스 함수를 사용하고, 손실 함수로는 크로스 엔트로피(categorical_crossentropy) 함수를 사용

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X_train, y_train, batch_size=128, epochs=5, verbose=1, validation_split=0.1)
    score = model.evaluate(X_test, y_test, batch_size=128, verbose=0)
    return score[1]

# 모델 설계를 fit_and_evaluate라는 함수 내에 정의하였는데, 모델을 함수 내에 정의한 이유는 이번 실습에서는 입력값을 바꿔가면서 모델을 여러번 호출하기 위함
```

```python
# 모델 훈련

modes = ['binary', 'count', 'tfidf', 'freq'] # 4개의 모드를 리스트에 저장.

for mode in modes: # 4개의 모드에 대해서 각각 아래의 작업을 반복한다.
    X_train, X_test, _ = prepare_data(train_email, test_email, mode) # 모드에 따라서 데이터를 전처리
    score = fit_and_evaluate(X_train, y_train, X_test, y_test) # 모델을 훈련하고 평가.
    print(mode+' 모드의 테스트 정확도:', score)
    
    
# 각 모드에 대해 총 5회의 에포크 수행
# 대체적으로 82% ~ 83%의 비슷한 정확도를 보이는데, 'freq' 모드에서만 정확도가 69%가 나옴
# 아무래도 'freq' 모드는 이번 문제를 풀기위한 적절한 전처리 방법이 아니었던 듯
```

# 피드 포워드 신경망 언어 모델 (NNLM, Neural Network Language Model)
< 정의 >
- 앞의 n개 단어를 참고하여 다음 단어 예측
    - n-gram 언어 모델과 유사

< 학습과정 >

- 1. 원-핫 인코딩
    - 훈련 코퍼스에 대해 원-핫 인코딩 수행
    - 원-핫 벡터들은 훈련을 위한 NNLM의 입력이면서, 예측을 위한 레이블(실제값 y)이 됨

- 2. Input Layer
    - 윈도우 (window) : 예측에 참고하는 앞의 단어의 개수 n의 범위
    
- 3. Projection Layer
    - 투사층 : 은닉층과 달리 활성화 함수가 존재하지 않는 선형층
    - 1. 룩업 테이블을 통해 원-핫 벡터 -> 임베딩 벡터로 변환
         - 원-핫 벡터와 가중치 W 행렬의 곱은 사실 W행렬의 i번째 행을 그대로 읽어오는 것과(lookup) 동일
         - 초기에는 랜덤한 값을 가지지만 학습 과정에서 값이 계속 변경됨
         - 워드 임베딩 (word embedding)
            - 단어 벡터 간 유사도를 구할 수 있는 벡터를 얻어내는 것
         
     - 2. concatenate
         - 모든 임베딩 벡터의 값 연결
         
         
- 4. Hidden Layer
    - 투시층의 결과는 크기가 h인 은닉층을 통과
    - 은닉층을 지난다는 것은 은닉층의 입력값에 가중치를 곱하고 편향을 더하여 그 값을 활성화 함수의 입력값으로 사용한다는 의미
    
    

- 5. Output Layer
    - 은닉층의 결과는 크기가 v인 출력층을 통과
    - 출력층을 지난다는 것은 출력층의 입력값에 가중치를 곱하고 편향을 더하여 그 값을 활성화 함수의 입력값으로 사용한다는 의미
    
    

- 6. 오차 비교
    - 손실함수로 cross-entropy를 사용하여 예측값y^와 실제값y의 오차 줄이기

- 7. 역전파
    - 모든 가중치 행렬이 학습되는 과정
        - 투사층에서의 가중치 행렬도 포함되므로 임베딩 벡터값 또한 학습됨

< 특징 >
- 장점 1
    - n_gram 언어 모델의 한계를 개선한 것
        - NNLM은 단어를 표현하기 위해 임베딩 벡터를 사용하므로서 단어의 유사도를 계산 가능 및 이를 통해 희소 문제(sparsity problem)를 해결
- 장점 2
    - 충분한 양의 훈련 코퍼스를 위와 같은 과정으로 학습한다면 결과적으로 수많은 문장에서 유사한 목적으로 사용되는 단어들은 결국 유사한 임베딩 벡터값을 얻게되는 것
    - 이렇게 되면 훈련이 끝난 후 다음 단어를 예측 과정에서 (마치 앞서 언급한 저자의 '톺아보기'와 같은 예시처럼) 훈련 코퍼스에서 없던 단어 시퀀스라 하더라도 다음 단어를 선택할 수 있음

- 한계
    - 고정된 길이의 입력(Fixed-length input)
         - n-gram 언어 모델과 마찬가지로 다음 단어를 예측하기 위해 모든 이전 단어를 참고하는 것이 아니라 정해진 n개의 단어만을 참고 가능
         - -> RNN으로 한계 극복 가능
