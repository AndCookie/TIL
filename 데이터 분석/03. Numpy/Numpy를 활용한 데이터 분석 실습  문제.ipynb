# [문제] BMI 지수 계산 및 과체중 이상인 BMI 값 확인
- 10명에 대한 키와 몸무게가 들어있는 'height_weight.txt'를 읽어 각 사람별 BMI 지수를 구하고 비만 전단계 이상인 BMI 값 확인하기

<pre>
 BMI 지수 = 몸무게(kg) / 키(M)의 제곱
 체질량지수의 범위 값에 따른 비만도 값 지정
 18.5 미만 ▶ 저체중
 18.5 ~ 22.9 ▶ 정상
 23 ~ 24.9 ▶ 비만 전단계
 25 ~ 29 ▶ 1단계 비만
 30 ~ 34.9 ▶ 2단계 비만
 35이상 ▶ 고도비만
</pre>

```python
import numpy as np

# np.loadtxt() 데이터를 불러오는 함수
# delimiter : 구분자 콤마,
# dtype : 데이터 타입
data = np.loadtxt('height_weight.txt', delimiter=',', dtype=float)
# print(data)
# print(type(data)) #ndarray

height = data[0]
weight = data[1]
# print(height)
# print(weight)

BMI = weight / ((height/100) ** 2)
print('각 사람별 BMI 지수: ', BMI)

# print(BMI>=23)
print('비만 전단계 이상인 BMI 값 확인하기: ', BMI[BMI>=23])
# 비만 전단계 이상인 BMI 값 확인하기
```

```python
# A.
weight = data[1]
print(weight)

# 키를 미터 단위로 바꾸고 제곱
squared_height_m = (data[0]/100) **2
print(squared_height_m)

bmi = weight / squared_height_m
print(bmi)

print(bmi[bmi>=23])
```

```python
# 여성과 남성의 수축기혈압과 이완기혈압의 평균값 확인


# 수축기혈압 : SBP
male_sbp_mean = np.mean(data[data[:,0]==1][:,2])
# 모든 데이터 중 0번째 열이 1인 것의 모든 데이터 중 2번째 열의 평균
print(f'남성의 수축기 평균 혈압: {male_sbp_mean:.2f}') # 소수점 두 자리

female_sbp_mean = np.mean(data[data[:,0]==2][:,2])
# 모든 데이터 중 0번째 열이 2인 것의 모든 데이터 중 2번째 열의 평균
print(f'여성의 수축기 평균 혈압: {female_sbp_mean:.2f}')


# 이완기혈압 : DBP
male_dbp_mean = np.mean(data[data[:,0]==1][:,3])
# 모든 데이터 중 0번째 열이 1인 것의 모든 데이터 중 3번째 열의 평균
print(f'남성의 이완기 평균 혈압: {male_dbp_mean:.2f}')

female_dbp_mean = np.mean(data[data[:,0]==2][:,3])
# 모든 데이터 중 0번째 열이 2인 것의 모든 데이터 중 3번째 열의 평균
print(f'여성의 이완기 평균 혈압: {female_dbp_mean:.2f}')
```

```python
# 정상 혈압 남자 및 여자 수 확인
# 정상혈압: 수축기 혈압 120nnHg 미만, 확장기 혈압 80mmHg 미만

nbp_masking = np.logical_and(data[:,2] < 120, data[:,3] < 80)
nbp = data[nbp_masking]

print(f'정상 혈압 남자 수: {sum(nbp[:,0]==1):,}')
print(f'정상 혈압 여자 수: {sum(nbp[:,0]==2):,}')
```

```python
# 당뇨병 질환(공복혈당 수치 126mg/dL 이상)을 가지고 있는 사람들의 평균 체질량지수
# 공복혈당 수치: 126 mg/dL 이상 -> 당뇨병, 100mg/dL-> 정상

disease = data[data[:,4] >= 126]
print(f'당뇨병 질환 환자 평균 BMI: {np.mean(disease[:,6]):.2f}')
```

# [문제] 시에틀 강수량 데이터 분석
- 'Seattle2014.csv' 파일 내용 읽기
- 강수량('PRCP')이 많은 날 상위 10일 확인
- 1월 강수량만 출력
- 1월에 내린 강수량의 합, 평균강수량 확인

```python
import numpy as np

data = np.loadtxt('Seattle2014.csv', delimiter=',', dtype=object, skiprows=1)
# numpy에서는 모든 데이터 타입이 동일해야하는데
# seattle에는 숫자와 문자 데이터가 혼합되어있다
# 따라서 모든 데이터를 문자열로 읽어오기 위해 dtype=object로 지정
print(type(data))
```

```python
# 강수량('PRCP')이 많은 날 상위 10일 확인
prcp = np.array(data[:,3], np.int32)
# print(prcp)
prcp_sort = np.sort(prcp)[::-1]
# print(prcp_sort)
masking = np.argsort(prcp_sort)[:10]
# print(masking)

print(f'강수량 상위 10일: {data[:,2][masking]}')
```

```python
# 1월 강수량만 출력
date = np.array(data[:,2], np.int64)
# print(date)
jan = date[0:31]
# print(jan)

print(f'1월 강수량: {prcp[0:31]}')
```

```python
# 1년동안 강수량이 40미만으로 내린 날이 몇일인가?

print(len(prcp[prcp < 40]))
```

```python
# A.

# 강수량('PRCP')이 많은 날 상위 10일 확인
prcp_data = data[:,3].astype(np.int32)
print(np.sort(prcp_data)[::-1][:10])

# 특정 컬럼을 기준으로 정렬하기
prcp_data = data[:,[2,3]].astype(int)
# print(prcp_data[:10])
print(prcp_data[prcp_data[:,1].argsort()][::-1][:10])
# prcp_data의 1번째 컬럼, 즉 강수량 데이터를 argsort하여 오름차순 정렬된 인덱스
```

```python
# A.

# 1월 강수량만 출력
prcp_data = data[:,[2,3]] # 문자열로 가져옴
# print(prcp_data[:5])

temp = np.array( [ [date[4:6], prcp] for date, prcp in prcp_data ] )
# print(temp[:5])

jan_prcp = temp [temp[:,0]=='01'][:,1].astype(int)
# boolean indexing : temp의 모든 레코드 중에서 첫번째 컬럼이 01인 것만 인덱싱
# [:,1] : 날짜값 버리고 강수량만 보겠다
# int로 형변환
print(jan_prcp)
```

```python
# A.

# 1월에 내린 강수량의 합, 평균강수량 확인
print(f'1월 강수량의 합: {np.sum(jan_prcp)}')
print(f'1월 평균 강수량: {np.mean(jan_prcp):.2f}')
```

# [문제] 영화 평점 분석
- 'ratings.dat' 파일 내용 읽기 (데이터 형식: user_id::item_id::rating::timestamp)
- 상위 10개 데이터 출력하기
- 데이터 정보 확인(크기, 차원)
- 전체 데이터 평점 구하기
- 사용자 아이디 1인 데이터 출력하기
- 사용자 아이디 1인 사람의 평점 평균 구하기
- 각 사용자별 평점 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
- 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기

```python
import numpy as np

data = np.loadtxt('ratings.dat', delimiter='::', dtype=int)
```

```python
# 상위 10개 데이터 출력하기

print(data[:10])
```

```python
# 데이터 정보 확인(크기, 차원)

print('데이터 크기: ', data.shape)
print('데이터 차원: ',data.ndim)
```

```python
# 전체 데이터 평점 평균구하기
print(f'전체 사용자 평균 평점: {np.mean(data[:,2]):.2f}')
```

```python
# 사용자 아이디 1인 데이터 출력하기
user_1 = data[data[:,0]==1]
print(user_1)
```

```python
# 사용자 아이디 1인 사람의 평점 평균 구하기

user_1 = data[data[:,0]==1]
print(np.mean(user_1[:,2]))
```

```python
# 각 사용자별 평점 평균 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
user_ids = np.unique(data[:,0])
print(f'전체 사용자 수: {len(user_ids):,}명')

for user, rate in data[:,[0,2]]:
    for unq_user in np.unique(data[:,0]):
        mn = np.mean(data[data[:,0]==unq_user][:,2] )
        print(f'{unq_user}번 평점 평균: {mn:.2f}')
```

```python
# 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기

print('사용자 ID: ', np.unique(data[:,0]))

high_user = []
for user, rate in data[:,[0,2]]:
    for unq_user in np.unique(data[:,0]):
        mn = np.mean(data[data[:,0]==unq_user][:,2])
        if mn >= 4.7:
            high_user.append(unq_user) 
    print(f'평균 평점이 4점 이상인 사용자 목록: {high_user}번')
#         print(f'평균 평점이 4점 이상인 사용자 인원수: {high_user}번')
```

```python
# A.

# 각 사용자별 평점 평균 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
user_ids = np.unique(data[:,0])
print(f'전체 사용자 수: {len(user_ids):,}명')

# 리스트 comprehension 이용하기
user_avg_rating = [ [id, np.mean(data[data[:,0]==id][:,2])] for id in user_ids ]
```

```python
# print(user_avg_rating[:5])
for id, rating in user_avg_rating[:10]:
    print(f'{id} : {rating:.2f}점')
```

```python
# A.
# 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기

# 파이썬 List -> 넘파이 ndarray로 바꿔주기
user_avg_rating = np.array(user_avg_rating)
print(f'평균 평점이 4점 이상인 유저 수: {np.sum(user_avg_rating[:,1] >= 4):,} 명')
```
